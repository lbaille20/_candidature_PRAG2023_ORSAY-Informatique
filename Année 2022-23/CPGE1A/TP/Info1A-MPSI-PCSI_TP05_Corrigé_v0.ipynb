## Exercice 1 (boucles `while`)

* rappel :

On redonne un exemple emblématique de boucle `while` réalisant le tirage d'un entier compris entre 1 et 6, jusqu'à obtenir un 6 :

```python
from random import randint

x = 0
while x < 6:
    x = randint(1, 6)
    print(x)
```

**Q1** Écrire une boucle `while` simulant le tirage aléatoire de deux entiers $a$ et $b$, compris entre 1 et 6, jusqu'à obtenir le couple $(a, b)=(6, 6)$. On tirera deux nouvelles valeurs tant que les deux valeurs tirées ne sont pas toutes deux égales à $6$. On affichera les valeurs de $a$ et $b$ à chaque tirage. 

***Q1 : corrigé***

```python
from random import randint

a, b = 0, 0 ## <--- on donne à a et b des valeurs arbitraires, de sorte que la condition d'entrée dans la boucle
            ## while soit vérifiée au moins une fois et qu'une itération au moins ait lieu
while (a, b) != (6, 6): ## <-- on répétera l'exécution du corps de la boucle tant que les deux valeurs tirées
                        ## ne sont pas toutes deux des 6
    a, b = randint(1, 6), randint(1, 6)
    print(a, b)
```

*Note :* la condition de &laquo;continuation&raquo; de la boucle `while` peut s'écrire aussi $(a, b) < (6, 6)$, compte-tenu de l'ordre défini sur les tuples, ou `not (a == 6 and b == 6)`, la dernière condition se réécrivant aussi `a != 6 or b != 6`, en vertu des lois de De Morgan.

* rappel des lois de De Morgan :

Les lois de De Morgan permettent de réécrire la négation d'une conjonction :
$${\rm NON} \left( \; A  \; {\rm ET}  \; b \; \right) = \left(\,{\rm NON} \; A \; \right) \;  {\rm OU} \;  \left(\,{\rm NON}  \; B  \; \right)$$
ou la négation d'une disjonction :
$${\rm NON} \left( \; A  \; {\rm OU}  \; b \; \right) = \left(\,{\rm NON} \; A \; \right) \;  {\rm ET} \;  \left(\,{\rm NON}  \; B  \; \right)$$

**Q2** Rependre la réponse à la question **1** en rajoutant un compteur permettant d'afficher après la fin de boucle, le nombre d'itérations nécessaires à l'obtention d'un couple $(6,6)$.

***Q2 : corrigé***

```python
from random import randint

a, b = 0, 0 
cpt = 0 ## <-- initialisation du compteur
while (a, b) != (6, 6):
    a, b = randint(1, 6), randint(1, 6)
    cpt = cpt + 1 ## <-- actualisation du compteur à chaque tirage d'un nouveau couple d'entiers
    print(a, b)
## après que la boucle while a terminé, on affiche le nombre de tirages qui ont dû être effectués
print("Nombre de tirages effectués :", cpt)
```

*Note :* si le tirage d'un entier entre 1 et 6 suit bien une loi uniforme sur l'intervalle d'entiers $[\,[1, 6]\,]$, les tirages de $a$ et $b$ étant indépendants, la probabilité d'obtenir le couple $(6, 6)$ est égale à $\frac{1}{6}\times\frac{1}{6}=\frac{1}{36} $ et le temps d'attente pour l'obtention du couple $(6,6)$ est une variable aléatoire d'espérance $36$.  
Le nombre moyen de tirages nécessaires à l'obtention du couple $(6,6)$ est donc égal à $36$.

On rappelle l'algorithme qui construit la copie *superficielle* d'une liste.  
Si la liste `L` en entrée est une liste de valeurs de type non-mutable (`int`, `float`, `str`, `bool` ou `tuple`), alors la copie construite est indépendante de la liste en entrée (*i.e.* une modification de `L` n'affecte pas la copie)

```python
def copie_liste(L):
    L1 = []
    for i in range(len(L)):
        L1.append(L[i])
    return L1
```

**Q2** Définir une fonction `copie_tranche(L, a, b)` prenant en argument une liste `L` et deux entiers $a$ et $b$ et renvoyant une nouvelle liste, constituée des éléments de `L` situés entre les positions $a$ (incluse) et $b$ (exclue) dans `L`.   
Si $b \leqslant a$, la fonction renverra une liste vide.  Les positions dans  correspondant à des indices de position non valides (`indexes out of range`) seront ignorées, ainsi que les valeurs strictement négatives.

***Q2 : corrigé***

Il s'agit de définir une fonction `copie_tranche(L, a, b)` qui, pour des valeurs positives de $a$ et $b$, renvoie une copie superficielle d'une portion de la liste `L` à l'identique de ce que produit l'instruction `L[a:b]`.<br/>
Pour des valeurs négatives de $a$ ou $b$, le comportement de la fonction est différent de celui du *slicing*.  
De façon générale, la fonction, telle qu'elle est définie renvoie la liste des valeurs de `L` pour les indices éléments de l'intersection $[\![a, b]\!]\cap[\![0, {\rm len(L)}-1]\!]$.

```python
def copie_tranche(L, a, b):
    if b <= a:
        return []
    ## si a <= b, on ignore les valeurs négatives :
    if a < 0:
        a = 0
    if b < 0:
        b = 0
    ## si a <= b, on ignore les indices > len(L) (indexes out of range) :
    if a > len(L):
        a = len(L)
    if b > len(L):
        b = len(L)
    # la condition de la boucle while permet d'ignorer les indices >= len(L) (indexes out of range)
    L1 = []
    i = a
    while i < b:
        L1.append(L[i])
        i += 1
    return L1

## tests
L = ['a', 'b', 'c', 'd', 'e', 'f']
print(L)
a, b = 0, 6
print('copie_tranche(L,', a, ',', b, ') :', copie_tranche(L, a, b))
a, b = 5, 3
print('copie_tranche(L,', a, ',', b, ') :', copie_tranche(L, a, b))
a, b = 3, 5
print('copie_tranche(L,', a, ',', b, ') :', copie_tranche(L, a, b))
a, b = 3, 12
print('copie_tranche(L,', a, ',', b, ') :', copie_tranche(L, a, b))
a, b = -2, 3
print('copie_tranche(L,', a, ',', b, ') :', copie_tranche(L, a, b))
a, b = -4, -2
print('copie_tranche(L,', a, ',', b, ') :', copie_tranche(L, a, b))
a, b = 6, 10
print('copie_tranche(L,', a, ',', b, ') :', copie_tranche(L, a, b))
```

```python
def copie_tranche2(L, a, b):
    a, b = max(0, a), min(len(L), b)
    L1 = []
    i = a
    while i < b:
        L1.append(L[i])
        i += 1
    return L1

## tests
L = ['a', 'b', 'c', 'd', 'e', 'f']
print(L)
a, b = 0, 6
print('copie_tranche2(L,', a, ',', b, ') :', copie_tranche2(L, a, b))
a, b = 5, 3
print('copie_tranche2(L,', a, ',', b, ') :', copie_tranche2(L, a, b))
a, b = 3, 5
print('copie_tranche2(L,', a, ',', b, ') :', copie_tranche2(L, a, b))
a, b = 3, 12
print('copie_tranche2(L,', a, ',', b, ') :', copie_tranche2(L, a, b))
a, b = -2, 3
print('copie_tranche2(L,', a, ',', b, ') :', copie_tranche2(L, a, b))
a, b = -4, -2
print('copie_tranche2(L,', a, ',', b, ') :', copie_tranche2(L, a, b))
a, b = 6, 10
print('copie_tranche2(L,', a, ',', b, ') :', copie_tranche2(L, a, b))
```

**Q4** Définir une fonction `plus_longue_suite_croissante`, prenant en argument une liste d'entiers `L`, supposée non vide, et renvoyant une liste composée des éléments `L[0]`, `L[1]`, $\dots$, `L[i]` de `L`, telle que $i$ est le plus grand entier positif tel que la suite finie `L[0]`, `L[1]`, $\dots$, `L[i]` soit définie et croissante.

***Q4 : corrigé***

Si ${\rm L} = [v_0, v_1, \ldots, v_{n-1}]$, la fonction doit renvoyer la liste $[v_0, v_1, \ldots, v_{i}]$ pour le plus grand entier $i$, $0\leqslant i< {\rm len(L)}$ tel que $v_0\leqslant v_1 \leqslant \ldots \leqslant v_i$.  
On remarque que, la liste `L` étant supposée non vide, la suite réduite au seul élément `L[0]` est croissante, et donc $i_0$ est toujours défini, et est au moins égal à $0$. 

Par exemple :<br/>
* `plus_longue_suite_croissante([2, 3, 1])` renverra `[2, 3]` ;  
* `plus_longue_suite_croissante([2, 3, 4, 5])` renverra `[2, 3, 4, 5]` ;  
* `plus_longue_suite_croissante([2, 1, 3])` renverra `[2]`.

Pour implémenter la fonction, on accumule dans une liste `L1`, initialisée à la liste `[L[0]]`, les valeurs `L[i]`, pour $i\geqslant 1$, tant que ${\rm L}[i] \geqslant {\rm L}[i-1]$. 

```python
def plus_longue_suite_croissante(L):
    L1 = [L[0]]
    i = 1
    while i < len(L):
        if L[i] >= L[i - 1]:
            L1.append(L[i])
        i += 1
    return L1
    
## tests
L = [2, 3, 1]
print(L, plus_longue_suite_croissante(L))
L = [2, 3, 4, 5]
print(L, plus_longue_suite_croissante(L))
L = [2, 1, 3]
print(L, plus_longue_suite_croissante(L))
```

**Q4** Définir une fonction `lettres_doubles`, prenant en argument un chaîne de caractères `ch` renvoyant `True` ou  `False`, selon qu'une chaîne contient au moins deux caractères consécutifs égaux ou non.  
Par exemple :  
* `lettres_doubles('abcd')` devra renvoyer le booléen `False` ;    
* `lettres_doubles('abcdd')` devra renvoyer le booléen `True` ;  
* `lettres_doubles('abcda')` devra renvoyer le booléen `False` ;      
* `lettres_doubles('aabcd')` devra renvoyer le booléen `True`.  

***Q4 : corrigé***

Il s'agit de parcourir la chaîne `ch` en renvoyant `True` dès que deux lettres consécutives sont égales.  
Si le parcours termine sans que `True` ait été renvoyée, la fonction doit renvoyer `False`. <br/> 
**Attention :** dans l'implémentation proposée, il faut veiller à ce qu'à chaque itération, à la fois `ch[i-1]` et `ch[i]` soient définis, *i.e.* $i$ doit être compris dans l'intervalle $[\![1, {\rm len(ch) -1}]\!]$, de sorte qu'à la fois $i-1$ et $i$ soient compris entre $0$ et $\rm len(ch) -1$.

```python
def lettres_doubles(ch):
    for i in range(1, len(ch)):
        if ch[i - 1] == ch[i]:  # <-- dès que deux lettres consécutives sont égales
            return True         # <-- le booléen True est renvoyé
    ## si le parcours termine sans que la valeur True ait été renvoyée
    return False # <-- le booléen False est renvoyé

## tests
ch = 'abcd'
res = lettres_doubles(ch)
print(ch, res)
ch = 'abcdd'
res = lettres_doubles(ch)
print(ch, res)
ch = 'abcda'
res = lettres_doubles(ch)
print(ch, res)
ch = 'aabcd'
res = lettres_doubles(ch)
print(ch, res)
```

La fonction peut aussi s'implémenter à l'aide d'une boucle `while`.  
Dans ce cas, on arrêtera le parcours, soit parce que l'on a rencontré deux lettres consécutives égales, soit par que l'on a atteint la fin de la chaîne.  
Lorsque la boucle termine, on renvoie `True` ou `False`, selon que pour la dernière valeur de $i$ testée, on a toruvé deux caractères consécutifs égaux ou non, ce qui est réalisé en renvoyant la valeur du test `ch[i - 1] == ch[i]`.  <br/>
**Important :** la condition `ch[i - 1] != ch[i]` ne doit pas être évaluée si $i = {\rm len(ch)} - 1$, donc la condition `i < len(ch) - 1`, doit être évaluée avant la condition `ch[i - 1] != ch[i]` de sorte que le test ne provoque pas une erreur `index out range` lorsque $i = {\rm len(ch)} - 1$.


```python
def lettres_doubles_while(ch):
    i = 1
    while i < len(ch) - 1 and ch[i - 1] != ch[i]: ## Attention à l'ordre des deux conditions 
        i += 1
    ## fin de la boucle while
    return ch[i - 1] == ch[i]

## tests
ch = 'abcd'
res = lettres_doubles_while(ch)
print(ch, res)
ch = 'abcdd'
res = lettres_doubles_while(ch)
print(ch, res)
ch = 'abcda'
res = lettres_doubles_while(ch)
print(ch, res)
ch = 'aabcd'
res = lettres_doubles_while(ch)
print(ch, res)
```

**Q5** Définir une fonction `est_palindrome`, prenant en argument un chaîne de caractères `ch` renvoyant `True` ou  `False`, selon qu'une chaîne se lit de la même façon, indifféremment, de gauche à droite ou de droite à gauche.  
"kayak" est un exemple de palindrome.

***Q5 : corrigé***

Une solution possible est de construire une chaîne `ch2` correspondant à la chaîne `ch` lue à l'envers,
et de renvoyer le résultat de la comparaison `ch2 == ch`, ce qui est fait ci-dessous, en utilisant les fonctions `renverser` et `solution1`.<br/>
La fonction `renverser` est construite un parcours de la liste `ch`, mais en construisant, caractère après caractère, une chaîne `ch2`, au lieu de simplement afficher les caractères lus.  
À chaque caractère lu, la chaîne `ch2` est recopiée et on lui ajoute un caractère supplémentaire.  
À l'itération d'indice $i$, on recopie les $i$ caractères de la chaîne `ch2` construite à l'itération précédente, afin de construire une nouvelle chaîne `ch2` augmentée du nouveau caractère lu, ce qui prend un temps proportionnel à $i$.<br/>
Ainsi on obtient une copie *à l'envers* de la chaîne `ch` avec une **complexité temporelle** (un coût en temps) proportionnel à $\sum_{i=0}^{{\rm len(ch)}-1} i= \frac{({\rm len(ch)}-1).{\rm len(ch)}}{2}$.  
Si on note $n = {\rm len(ch)}$, pour $n$ grand, le temps d'exécution de `renverser(ch)` est équivalent à $\frac{1}{2}n^2$.  
S'ajoute à ce temps, le temps de comparer les chaînes `ch` et `renverser(ch)`, ce qui coûte un temps au pire proportionnel à la longueur de la chaîne `ch` (dans le cas où `ch` est bien un palindrome, il faut comparer tous les caractères des deux chaînes pour constater leur égalité).

```python
def renverser(ch):
    ch2 = ""
    for i in range(len(ch)):
        ch2 = ch[i] + ch2
    return ch2

## test
renverser("0123456789")
```

```python
def solution1(ch):
    return renverser(ch) == ch

## tests
ch = 'kayak'
res = solution1(ch)
print(ch, res)
ch = 'abcd'
res = solution1(ch)
print(ch, res)
```

**On préférera de loin retenir la solution suivante, dont le temps d'exécution est au pire proportionnel à la moitiée de la longueur de la chaîne `ch`.**

Cette solution consiste à parcourir la première moitié de la chaîne `ch` pour comparer chaque caractère à son correspondant dans la seconde moitié de la chaîne (son symétrique par rapport au milieu de la chaîne), en interrompant le parcours dès qu'un caractère est différent de son correspondant.

```python
def est_palindrome(ch):
    n = len(ch)
    for i in range(n // 2):
        if ch[i] != ch[n - 1 - i]:
            return False
    return True

## tests
## tests
ch = 'kayak'
res = est_palindrome(ch)
print(ch, res)
ch = 'abcd'
res = est_palindrome(ch)
print(ch, res)
```

On peut mesurer le temps d'exécution à l'aide de la fonction `perf_counter` du module `time` qui permet, en calculant la différence entre deux valeurs renvoyées par un appel `perf_counter()`, d'obtenir le temps, en secondes, écoulé pour exécuter un appel à la fonction testée.  
On pourra tester les deux fonctions pour des chaînes de longeurs, 10, 100 et 1000 caractères, qui sont ou non des palindromes.

```python
from time import perf_counter

def affiche_temps_execution(ch, n):
    t1 = perf_counter()
    res = solution1(ch)
    t2 = perf_counter()
    d = t2 - t1
    print("n =", n, "solution1 renvoie :", res, "temps d'exécution :", d)
    t1 = perf_counter()
    res = est_palindrome(ch)
    t2 = perf_counter()
    d = t2 - t1
    print("n =", n, "est_palindrome renvoie :", res, "temps d'exécution :", d)
```

```python
## tests
for n in (10, 100, 1000, 10**4, 10**5, 10**6):
    ch1 = n * 'a' ## palindrome
    affiche_temps_execution(ch1, n)
```

On peut constater que pour une chaîne qui est un palindrome, entre les valeurs de $n$ $10^5$ et $10^6$, le temps d'exécution pour la fonction `est_palindrome` est multiplié par environ 10 et par environ 100 pour la fonction `solution1`.

```python
## tests
for n in (10, 100, 1000, 10**4, 10**5, 10**6):
    ch2 = (n // 2) * 'a' + (n // 2) * 'b' ## non-palindrome
    affiche_temps_execution(ch2, n)
```

On peut constater que pour une chaîne qui n'est pas un palindrome (par simple examen de ces premier et dernier caractères), entre les valeurs de $n$ $10^5$ et $10^6$, le temps d'exécution pour la fonction `est_palindrome` peut-être considéré comme constant et est multiplié par environ 100 pour la fonction `solution1`.

Ainsi, les chaînes `""` (chaîne vide), `"a"`, `"abr"`, `"abracadra"` sont des préfixes de la chaîne `"abracadabra"`(de longueur respectives $0$, $1$, $3$ et $11$).  

**Q1** Écrire une fonction `prefixe(ch, m)` renvoyant le préfixe de longueur $m$ de la chaîne `ch`.  
La fonction renverra `None` s'il n'existe pas de préfixe de longueur $m$ pour la chaîne `ch`.

***Q1 : corrigé***

Pour $m$ compris entre $1$ et $\rm len(ch)$, il s'agit de copier les $m$ premiers caractères de la chaîne `ch` et de renvoyer la chaîne ainsi obtenue.<br/>
Pour les autres valeurs de $m$, la fonction doit renvoyer une chaîne vide.

Voici une solution n'utilisant pas le *slicing*, ce qui n'est pas la solution attendue :

```python
def prefixe0(ch, m):
    if m < 0 or m > len(ch):
        return None
    ## si on ne renvoie pas la chaîne vide, on copie les m premiers caractères de ch dans une chaîne pref
    ## que l'on renvoie ensuite
    pref = ""
    for i in range(m):
        pref = pref + ch[i]
    return pref

# tests
ch, m = "abracadabra", 0
res = prefixe0(ch, m)
print("ch :", ch, "; préfixe de longueur", m, ":", res)
ch, m = "abracadabra", 5
res = prefixe0(ch, m)
print("ch :", ch, "; préfixe de longueur", m, ":", res)
ch, m = "abracadabra", 16
res = prefixe0(ch, m)
print("ch :", ch, "; préfixe de longueur", m, ":", res)
```

La réponse attendu utilise le *slicing* et la voici :

```python
def prefixe(ch, m):
    if m <= len(ch):
        return ch[0:m]

# tests
ch, m = "abracadabra", 0
res = prefixe(ch, m)
print("ch :", ch, "; préfixe de longueur", m, ":", res)
ch, m = "abracadabra", 5
res = prefixe(ch, m)
print("ch :", ch, "; préfixe de longueur", m, ":", res)
ch, m = "abracadabra", 16
res = prefixe(ch, m)
print("ch :", ch, "; préfixe de longueur", m, ":", res)
```

On parcourt les deux chaînes simultanément, dès que les lettres de même indice de chaque chaîne ne correspondent pas, on renvoie les caractères communs d'indice strictement inférieur.  
Afin de ne pas tester des caractères à des positions non définies, on borne la recherche aux caractères d'indice inférieurs strictement au minimum entre ${\rm len(ch1)}$ et ${\rm len(ch2)}$.

* s'il est interdit de recourir à la fonction `min` :

```python
def plus_long_prefixe_commun(ch1, ch2):
    if len(ch2) < len(ch1):
        m = len(ch2)
    else:
        m = len(ch1)
    for i in range(m):
        if ch1[i] != ch2[i]:
            return ch1[0:i]
    # si la boucle for termine sans qu'un préfixe ait été renvoyé, on renvoie la plus courte des deux chaînes
    if len(ch2) < len(ch1):
        return ch2
    else:
        return ch1
    
## tests
ch1, ch2 = "abcd", "efgh"
res = plus_long_prefixe_commun(ch1, ch2)
print(ch1, ch2, "; plus long préfixe commun :", res)
ch1, ch2 = "abcd", "abcdefgh"
res = plus_long_prefixe_commun(ch1, ch2)
print(ch1, ch2, "; plus long préfixe commun :", res)
ch1, ch2 = "python", "py"
res = plus_long_prefixe_commun(ch1, ch2)
print(ch1, ch2, "; plus long préfixe commun :", res)
ch1, ch2 = "abcij", "abckl"
res = plus_long_prefixe_commun(ch1, ch2)
print(ch1, ch2, "; plus long préfixe commun :", res)
```

**Q3** Modifier la fonction `position_premiere_occurrence1` en une fonction `recherche_motif1(texte, motif)` renvoyant la position de la première occurrence de la chaîne `motif` au sein de la chaîne `texte`.  
La position d'une occurrence sera donnée par la position du premier caractère de la chaîne `motif`, là où on la trouve dans la chaîne `texte`.  
La fonction renverra $-1$ si le motif n'est pas présent dans la chaîne.  
On utilisera le *slicing* pour extraire des tranches de la chaîne `texte` afin de les comparer à la chaîne `motif`.  

***Q3 : corrigé***

Il suffit de changer le test `ch[i] == c` en le test `texte[i:i + M] == motif` où $M$ est la longueur du motif.  
Afin d'éviter des tests inutiles, on observera que `texte[i:i + M]` définit une tranche de longueur $M$, seulement si, $0\leqslant i < i + M\leqslant {\rm len(texte)}$.<br/>
Si $i + M > {\rm len(texte)}$, *i.e.* $i > {\rm len(ch)} - M$, la tranche `texte[i:i + M]` est de longueur strictement inférieure à $M$ et donc il est inutile d'effectuer le test `texte[i:i + M] == motif`.<br/>
En conséquence, on limitera les valeurs de $i$ testées à celles couvertes par `range(0, len(texte) - M + 1)`.

```python
def recherche_motif1(texte, motif):
    T, M = len(texte), len(motif)
    for i in range(T - M + 1):
        if texte[i:i + M] == motif:
            return i
    # si le motif n'a pas été trouvé
    return -1

## tests
texte, motif = "tintin", "tin"
res = recherche_motif1(texte, motif)
print(texte, "; motif :", motif, "; résultat de recherche :", res)
texte, motif = "Tintin", "tin"
res = recherche_motif1(texte, motif)
print(texte, "; motif :", motif, "; résultat de recherche :", res)
texte, motif = "tonton", "tin"
res = recherche_motif1(texte, motif)
print(texte, "; motif :", motif, "; résultat de recherche :", res)
```

**Q4** Donner une seconde version, `recherche_motif2`, de la fonction précédente, en remplaçant la comparaison d'une tranche de texte au motif, déléguée à Python par appel à l'opérateur "==", par une comparaison, caractère par caractère, des caractères de la tranche et de ceux du motif. On recourra pour cela à une boucle `while`.<br/>
On pourra s'inspirer de la fonction `plus_long_prefixe_commun` de l'**exercice 5**.

***Q4 : corrigé***

```python
def recherche_motif2(texte, motif):
    T, M = len(texte), len(motif)
    for i in range(T - M + 1):
        ## on parcourt simultanément la tranche texte[i: i + M] et le motif à l'aide d'une boucle while
        ## d'indice j. La valeur de j s'interprète comme une position au sein du motif, tandis que
        ## la valeur i + j s'interprète comme une position au sein de la chaîne texte,
        ## comprise entre i et i + M - 1
        j = 0
        while j < M and texte[i + j] == motif[j]:
            j = j + 1
        ## lorsque la boucle while termine, le motif a été trouvé si et seulement si j == M
        ## auquel cas on peut renvoyer la valeur i qui donne la position d'une première occurrence du motif
        ## dans la chaîne  texte
        if j == M:
            return i
    # si le motif n'a pas été trouvé
    return -1

## tests
texte, motif = "tintin", "tin"
res = recherche_motif2(texte, motif)
print(texte, "; motif :", motif, "; résultat de recherche :", res)
texte, motif = "Tintin", "tin"
res = recherche_motif2(texte, motif)
print(texte, "; motif :", motif, "; résultat de recherche :", res)
texte, motif = "tonton", "tin"
res = recherche_motif2(texte, motif)
print(texte, "; motif :", motif, "; résultat de recherche :", res)
```
