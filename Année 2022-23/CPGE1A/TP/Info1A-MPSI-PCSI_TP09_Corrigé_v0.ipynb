***Q1.c corrigé***

On demande ici une ***proposition*** de majoration dans le sens où aucune démonstration n'est attendue.  
Dans le corrigé de cette question, la démonstration par récurrence de la majoration **( * )** n'est pas présentée (on trouvera une démonstration analogue, faite, dans les éléments de cours sur la recherche dichotomique dans une liste triée).  
Il est également admis **( ** )** que la valeur à trouver est nécessairement dans l'intervalle de recherche délimité par les valeurs des variables `a` et `b`.

* Majoration du nombre d'essais pour des valeurs de $N$ de la forme $2^m$ (donc lorsque $N$ est égal à une puissance de $2$) :  
  
Si $N$ est de la forme $2^m$, le nombre de valeurs initialement dans l'intervalle $[\![1;N]\!]$ est égal à $2^m$, et après une proposition, l'intervalle contient (au plus) $\frac{2^m}{2}=2^{m-1}$ valeurs.  
On démontre **(*)** alors, par récurrence, qu'après $p$ propositions, le nombre de valeurs dans l'intervalle de recherche est, au plus, égal à $\frac{2^m}{2^p}=2^{m-p}$.  
Ainsi, la recherche aura nécessairement terminé dès que l'on aura fait un nombre de propositions au plus égal à $p_N$, où $p_N$ est le plus petit entier naturel $p$ tel que $2^{m-p}\leqslant 1$.  En effet, après $p_N$ propositions, l'intervalle de recherche contiendra nécessairement au plus une valeur, valeur qui sera proposée et sera nécessairement égale à la valeur à trouver, puisqu'il est assuré **( ** )** que la valeur à trouver est dans l'intervalle de recherche.  
La résolution de l'inéquation d'inconnue $p$, $2^{m-p}\leqslant1$, se ramène aux équivalences suivantes :  
$$
\left(2^{m-p}\leqslant 1\right)
\Longleftrightarrow \left(2^{m}\leqslant 2^{p}\right) 
\Longleftrightarrow \left(m\leqslant p\right).
$$
On en conclut que la recherche aura terminé après au plus $p_N=m$ propositions.

* Majoration du nombre d'essais pour une valeur quelconque de $N$ :  
  
En continuant d'admettre qu'avant toute itération et à la fin de chaque itération de la boucle `while`, la valeur cherchée, `x` est dans l'intervalle de recherche d'extrémités `a` et `b`, en admettant encore que le nombre d'entiers dans l'intervalle de recherche diminue de moitié à chaque itération tant qu'il n'est pas réduit à un seul élément au plus, alors, si on note $m_N$ le plus petit entier tel que $n\leqslant 2^{m_N}$, alors le raisonnement précédent donne que la recherche aura terminé après au plus $p_N=m_N$ itérations.  
  
Sachant que $N>0$, la résolution de l'inéquation d'inconnue $m$, $N\leqslant 2^{m}$, se ramène aux équivalences suivantes :  
$$
\left( N\leqslant 2^{m} \right)
\Longleftrightarrow \left( \ln N \leqslant \ln 2^m \right) 
\Longleftrightarrow \left( \ln N \leqslant m \ln 2 \right) 
\Longleftrightarrow \left( \frac{\ln N}{\ln 2} \leqslant m \right) 
\Longleftrightarrow \left( \log_2 N \leqslant m \right) 
\Longleftrightarrow \left( m \geqslant \lceil \log_2 N \rceil \right).
$$
où $\lceil \log_2 N \rceil$ désigne la partie entière supérieure de $\log_2 N$ (logarithme en base $2$ de $N$, dont une valeur approchée est obtenue par appel aux fonctions `log2` - logarithme en base 2, et `ceil` - partie entière supérieure, du module `math`).  

On en conclut que la recherche aura terminé après au plus $p_N = \lceil \log_2 N \rceil$ propositions.

* tests

Afin de confronter ces résultats à l'expérimentation, on propose la fonction suivante, `teste_algo`, implémentant la recherche d'une suite optimale de propositions pour une valeur à trouver choisie au hasard entre dans un intervalle $[\![1;N]\!]$, en suivant l'algorithme proposé.  
La fonction `teste_algo` prend en paramètre un entier $N \geqslant 1$, affiche les extrémités successives des intervalles de recherche et leurs nombres d'éléments entiers, et renvoie la liste des propositions faites, ainsi que leur nombre et la valeur de la borne supérieure estimée pour le nombre de propositions à faire.

```python
def teste_algo(N):
    from random import randint
    from math import log2, ceil

    x = randint(1, N)
    m = ceil(log2(N))
    print('Valeur à trouver :', x)
    print('Nombre maximal d\'essais estimé :', m)
    y = 0 
    a, b = 1, N 
    L = []
    while y != x:
        print(f'intervalle de recherche :  [{a}, {b}] (nombre d\'éléments : {b - a + 1})')
        y = (a + b) // 2
        L.append(y)
        if y < x:
            a = y + 1
        elif x < y:
            b = y - 1
    print("Dernière valeur proposée :", y)
    print("Valeur trouvée avec", len(L), "propositions.")
    return L, len(L), m
```

* Nombre de propositions pour toutes les valeurs de $x$ entre $1$ et $N$ pour $N=100$ :

***À titre d'illustration (non demandé en TP)***, pour, par exemple, $N=100$, on peut afficher le nombre de propositions faites en appliquant l'algorithme proposé, pour toutes les valeurs $x$ à trouver possibles, comprises entre $1$ et $N=100$,  en se dotant de la fonction `trouve_algo(N, x)` implémentant la recherche dichotomique d'une valeur $x$ comprise entre $1$ et un entier $N\geqslant 1$ donné, et renvoyant le nombre de propositions faites :

```python
def trouve_algo(N, x):
    y = 0 
    a, b = 1, N 
    compteur = 0
    while y != x:
        y = (a + b) // 2
        compteur += 1
        if y < x:
            a = y + 1
        elif x < y:
            b = y - 1
    return compteur
```

On construit alors, pour $N\geqslant 1$ donné, ici $N=100$, la liste du nombre de propositions faites pour trouver chaque valeur $x$ comprise entre $1$ et $N$ :

```python
N = 100
Lresultats = []
for x in range(1, N + 1):
    Lresultats.append(trouve_algo(N, x))
```

On peut alors obtenir le tracé des nombre de propositions en fonction de la valeur de $x$ :

```python
import matplotlib.pyplot as plt

N = 100

fig, ax = plt.subplots()

ax.plot([x for x in range(1, N + 1)], [trouve_algo(N, x) for x in range(1, N + 1)])
ax.set_title("Nombre de propositions pour tout $x$ compris entre $1$ et $N=100$")
ax.set_xlabel('$x$')
ax.set_ylabel('nombre de propositions')
plt.show()
```

## Exercice 2 &emsp; exponentiation rapide

L'algorithme d'exponentiation rapide cherche à minimiser le nombre de multiplications nécessaires pour calculer une puissance $n^{ème}$.  
Pour tout $n$ entier naturel, le calcul de $x^n$ nécessite $n-1$ multiplication(s), si l'on considère l'égalité $x^n= x\times x\times \ldots \times x$.  
Le temps d'exécution de ces $n-1$ multiplication(s) n'est pas un obstacle pour de petites valeurs de $n$, mais la minimisation du nombre de multiplications peut s'avérer nécessaire pour de très grandes valeurs de $n$, ou si $x$ désigne une matrice.

**Q1** Pour calculer $x^8$, est-il possible d'effectuer moins de $7$ multiplications ? Comment procéder ?  
La méthode proposée s'adapte-t-elle pour calculer $x^{13}$ ?

***Q1 corrigé***

On peut calculer $x^8$ en écrivant $x^8=\left(\left(\left(x\right)^2\right)^2\right)^2$. Avec cette décomposition, on calcule $x^8$ en calculant trois carrés, ce qui nécessite trois multiplications (une pour chaque carré).  
On peut généraliser immédiatement la méthode pour le calcul de $x^{2^p}$ pour tout $p\geqslant 0$, nécessitant $p$ calculs de carrés, donc $p$ multiplications (au lieu de $2^p-1$ multiplications en suivant la définition d'une puissance entière positive d'un nombre $x$).

La réécriture utilisée pour calculer $x^8$ ne vaut que si l'exposant est une puissance de $2$, ce qui n'est pas le cas si l'on doit calculer $x^{13}$.  
Cependant, de façon générale, pour le calcul de $x^n$, $n\geqslant 1$, dans tous les cas, on peut néanmoins :
* si $n$ est pair, considérer l'écriture de $n$ sous la forme $m=2p$, car avec ces notations, il vient :
$$x^n=x^{2p}=\left(x^p\right)^2,$$
ce qui immédiatement, sachant qu'il faut au maximum $p-1$ multiplications pour calculer $x^p$, ramène les $n-1$ multiplications pour calculer $x^n$ en suivant la définition, à au plus $(p-1)+1$ multiplications ($p-1$ au maximum pour calculer $x^p$ et une pour en calculer le carré) ;    
* si $n$ est impair, considérer l'écriture de $n$ sous la forme $m=2p+1$, et, avec ces notations, il vient :
$$x^n=x^{2p+1}=x.\left(x^p\right)^2,$$
ce qui immédiatement, sachant qu'il faut au maximum $p-1$ multiplications pour calculer $x^p$, ramène les $n-1$ multiplications pour calculer $x^n$ en suivant la définition, à au plus $(p-1)+2$ multiplications ($p-1$ au maximum pour calculer $x^p$ et une pour en calculer le carré, et une pour multiplier le résultat par $x$).

Avec cette méthode, on réécrira $x^{13}$ sous la forme $x^{13}=x^{12+1}=x^{1+2\times 6}=x.\left(x^6\right)^2$.  
Si l'on itère la méthode, on calculera $x^6$ par la réécriture $x^6=x^{2\times 3}=x.\left(x^3\right)^2$, puis $x^3$ par la réécriture $x^3=x.x^2$.  
En conclusion, on aura calculé $x^{13}$ avec la décomposition suivante :
$$
x^{13} = x.\left( \left( x . \left( x \right)^2 \right)^2\right)^2
,
$$
ce qui nécessite deux multiplications par $x$, et trois calculs de carrés, soit au total $2+3=5$ multiplications au lieu de $13-1=12$ en suivant la définition.

***Q3 corrigé***

L'implémentation itérative de l'algorithme précédent n'est pas immédiate.  
On peut s'appuyer, pour la mise en place du calcul de $x^n$, sur la division euclidienne de $n$ par $2$, écrite sous la forme :
$$n = 2q + r, \; \text{ où } \; 0\leqslant r< 2.$$
Le calcul de $x^n$ par la méthode précédente consiste alors à effectuer le calcul en suivant la décomposition 
$$x^n=x^{2q+r}=x^r.\left(x^q\right)^2.$$
Si $q=0$, on a $x^q=1$, et pour $q>0$, si l'implémentation est récursive, on peut déléguer le calcul de $x^q$ à un appel récursif, mais si l'implémentation est itérative, on ne peut qu'itérer pour calculer la valeur de $x^q$. Il s'agirait donc d'itérer pour calculer $x^q$, ce qui nécessite de mettre en attente la multiplication par $x$ si $r=1$, et de mettre en attente le calcul du carré de $x^q$.   

Une solution consiste alors à utiliser une variable auxiliaire, `v_aux` dans l'implémentation ci-dessous, initialisée à 1, pour stocker les multiplications en attente, et à pousuivre les itérations pour calculer $\left(x^q\right)^2$, que l'on peut écrire sous la forme $\left(x^2\right)^q$ afin d'utiliser, comme pour la décomposition de $x^n$, une division euclidienne pour décomposer le calcul de la puissance $q^{ème}$ de $x^2$, et éviter ainsi de mettre en attente le calcul du carré de $x^q$.  
Afin de répéter la décomposition grâce à une division euclidienne par $2$, on utilisera une variable auxiliaire, `exposant`, initialement égale à $n$ et actualisée à la valeur du quotient dans la division euclidienne par 2 de l'exposant de la puissance à calculer à chaque itération.  
Il conviendra en parallèle de maintenir une variable auxilaire, `x_aux`, initialisée à la valeur de $x$ et élevée au carré à la fin de chaque itération.

```python
def expo_rapide_it(x, n):
    v_aux, exposant = 1, n
    x_aux = x
    while exposant > 0:
        q, r = exposant // 2, exposant % 2
        if r == 1:
            v_aux = v_aux * x_aux
        exposant = q
        x_aux = x_aux * x_aux
    return v_aux

## tests
x, n = 3, 8
r, r1 = x ** n, expo_rapide_it(x, n)
print(r, r1, r == r1)

x, n = 2.5, 13
r, r1 = x ** n, expo_rapide_it(x, n)
print(r, r1, r == r1)
```

Posons, pour $k=0$, $v_k=v_0=1$, $x_k=x_0=x$ et $p_k=p_0=n$,    
et, pour tout $k\geqslant 0$, tel que $p_k>0$, notons $p_k=2q_k+r_k$ la division euclidienne de $p_k$ par $2$, et posons :
* $v_{k+1} = v_k.\left(x_k\right)^{r_k}$, soit, explicitement, $v_{k+1}=v_k \times x_k$ si $r_k=1$, *i.e.* si $p_k$ est impair, et $v_{k+1}=v_k$ sinon ;
* $x_{k+1} = x_k \times x_k = \left(x_k\right)^2$ ;  
* $p_{k+1} = q_k$.

Avec ces notations, $v_0$, $x_0$, $p_0$ sont les valeurs respectives des variables `v_aux`, `x_aux` et `exposant` avant toute itération, et  
$v_k$, $x_k$, $p_k$ les valeurs respectives de ces mêmes variables à la fin de l'itération numéro $k$, pour tout $k\geqslant 1$, si elle a lieu.

Avec ces notations, on a encore :
* pour $k=0$, $v_k \times \left( x_k\right)^{p_k}= v_0 \times \left( x_0\right)^{p_0}
=1 \times \left( x\right)^{n}=x^n$ ;
* et, pour tout $k\geqslant 0$ tel que $p_k>0$, si $v_k \times \left( x_k\right)^{p_k}=x^n$ :
$$
v_{k+1} \times \left( x_{k+1}\right)^{p_{k+1}}=
v_k.\left(x_k\right)^{2q_k+r_k}=
v_k.\left(x_k\right)^{p_k}=
x^n.
$$

Ceci permet d'établir, par récurrence, qu'initialement et à la fin de toute itération d'indice $k$, $k\geqslant 1$, si elle a lieu, la propriété 
$$\left({\mathcal Q_k}\right) \quad
\left(v_k \times \left( x_k\right)^{p_k}=x^n\right)$$
est vraie.  
La propriété $\left({\mathcal Q_k}\right)$ est ainsi un **invariant de boucle** pour l'algorithme implémenté de façon itérative au travers de la fonction `expo_rapide_it`, ***assurant que, lorsque la boucle `while` termine***, *i.e.* lorsque la variable `exposant` prend la valeur zéro, ou encore pour le plus petit $k$ tel que $p_k=0$, on a :
$$v_k \times \left( x_k\right)^{p_k}= v_k \times \left( x_k\right)^{0}=v_k =x^n.$$

**Q4** On donne l'algorithme suivant permettant d'obtenir l'écriture binaire d'un entier $n$ strictement positif :

```python
def ecriture_binaire(n):
    ch = ""
    q = n
    while q > 0:
        q, r = q // 2, q % 2
        ch = str(r) + ch
    return ch
```

**Q5** On rappelle que l'écriture binaire d'un entier $n>0$ est donnée par $\left(c_{m-1}\ldots c_1c_0\right)_2$, où $m$ est l'unique entier naturel $m$ tel qu'il existe une suite, $c_{m-1}, \ldots,  c_1, c_0$,  de $m$ entiers de $\left\{0, 1\right\}$ tels que $c_{m-1}=1$ et 
$$
n = \sum_{k=0}^{m-1}c_k.2^k.
$$
On notera bien que la valeur de l'entier $m$ donne le nombre de chiffres de l'écriture binaire de $n$.

**Q5.a** Donner, en fonction de $m$,  un encadrement des entiers $n>0$ dont l'écriture binaire comporte exactement $m$ chiffres.

***Q5.a corrigé***

Les entiers $n>0$ dont l'écriture comporte exactement $m$ chiffres ($m\geqslant 1$) sont les entiers dont l'écriture binaire s'écrit, puisqu'alors $c_{m-1}=1$ :
$$
n = 2^{m-1}  + \sum_{k=0}^{m-2}c_k.2^k.
$$
Comme les chiffres $c_0, c_1, \ldots c_{m-2}$ de cette écriture sont, chacun, au plus bas, égaux à zéro, et, au plus haut, égaux à un, on en déduit que les entiers $n$ cherchés sont des entiers vérifiant :
$$
2^{m-1}  + \sum_{k=0}^{m-2}0.2^k
\leqslant n \leqslant 
2^{m-1}  + \sum_{k=0}^{m-2}1.2^k
$$
*i.e.* les entiers $n$ tels que :
$$
2^{m-1}
\leqslant n \leqslant 
\sum_{k=0}^{m-1}2^k.
$$
Comme la somme $\sum_{k=0}^{m-1}2^k$ vaut $1.\frac{1-2^{(m-1)-0 + 1}}{1-2}$ (somme des termes d'indices $0$ à $m-1$ de la suite géométrique de premier terme $1$ et de raison $q=2$) *i.e.* vaut, après simplification, $2^{}-1$, on en déduit que les entiers cherchés sont tels que :
$$
\left(2^{m-1}
\leqslant n \leqslant 
2^{m}-1\right)\Longleftrightarrow
\left(2^{m-1}
\leqslant n < 
2^{m}\right),
$$
où l'équivalence vaut car $n$ est entier.

Comme l'intervalle d'entiers $\left[\!\left[2^{m-1};2^{m}-1\right]\!\right]$ contient $(2^{m}-1)-2^{m-1}+1=2^{m}-2^{m-1}=2^{m-1}(2-1)=2^{m-1}$ valeurs, et qu'il existe $2^{m-1}$ $m$-uplets $(c_{m-1}, c_{m-2},\ldots, c_1, c_0)$ de la forme $(1, c_{m-2},\ldots, c_1, c_0)$ où les $c_i$, $0\leqslant i\leqslant m-2$ sont compris entre $0$ et $1$, on en conclut que :  
<br/>
<div style="text-align:center">les entiers $n>0$ dont l'écriture binaire comporte exactement $m$ chiffres sont exactement les entiers de l'intervalle $\left[\!\left[2^{m-1};2^{m}-1\right]\!\right]$.</div>

Conclure en donnant la valeur du nombre de chiffres de l'écriture en base 2 d'un entier $n>0$ à l'aide d'une partie entière.

\left(2^{m-1}

2^{m}-1\right)
\Longleftrightarrow
\left(\log_2\left(2^{m-1}\right)
\leqslant \log_2 n < 
\log_2\left(2^{m}\right)\right)
\Longleftrightarrow

m.\log_2 2\right)

On déduit du dernier encadrement, et de la définition de la partie entière d'un réel $x$, comme étant l'unique entier $a$ tel que $a\leqslant x< a +1$, noté $\lfloor x\rfloor$, que le nombre de chiffres de l'écriture en base 2 d'un entier $n>0$ est l'entier $m$ tel que $m-1=\lfloor \log_2 n\rfloor$, *i.e.* :

**Q5.c** Déduire de l'examen des algorithmes implémentés aux questions **Q3** et **Q4** et de la question **Q5.b** le nombre d'itérations nécessaires au calcul d'une puissance $n^{ème}$ à l'aide de l'algorithme d'exponentiation rapide.

***Q5.c corrigé***

En comparant les algorithmes implémentés par les fonctions `expo_rapide_it(x, n)` et `ecriture_binaire(n)`, on constate que, pour une même valeur de $n$, les valeurs prises dans l'un par la variable `exposant` et dans l'autre par la variable `q` sont les mêmes - et qu'en conséquence le nombre d'itérations de la boucle `while` est le même, tandis que les variables `r` y prennent les mêmes valeurs.  
    On en déduit que le nombre d'itérations de la boucle `while` est égale au nombre de chiffres de l'écriture binaire de $n$ (chaque itération de la boucle `while` dans la fonction de `ecriture_binaire(n)` produisant un chiffre de l'écriture binaire de $n$) et que le nombre de multiplications pour calculer $x^n$ est égal au nombre de ces itérations (on dénombre là le nombre de multiplications pour l'actualisation de la valeur de `x_aux` *via* l'instruction `x_aux = x_aux * x_aux`) auquel on ajoute le nombre de chiffres égaux à 1 dans l'écriture binaire de $n$ (on dénombre là le nombre de multiplications pour l'actualisation de la valeur de `v_aux` *via* l'instruction `v_aux = v_aux * x_aux`).  
Ainsi :
<br/>
<div style="text-align:center">le nombre de multiplicatios pour calculer la puissance $n^{ème}$ d'un nombre $x$ est exactement égal au nombre de chiffres de l'écriture en base 2 de $n$ augmenté du nombre de chiffres égaux à 1 dans cette écriture.</div>
De façon générale, si $m$ est le nombre de chiffres de l'écriture en base 2 de $n$, ce nombre de multiplication, noté ici $N_n$,  est donc compris entre $m+1$ (il y a au minimum un chiffre égal à 1 dans l'écriture binaire de $n$ - son bit de poids fort) et $2m$ (lorsque tous les chiffres de l'écriture binaire de $n$ sont égaux à 1).  
   
On peut donc donner en conclusion l'encadrement suivant du nombre de multiplications nécessaire avec l'algorithme d'exponentiation rapide dans son implémentation itérative :
$$
\lfloor \log_2 n\rfloor + 1\leqslant N_n \leqslant 2.\left(\lfloor \log_2 n\rfloor + 1\right).
$$

Sur les deux exemples considérés précédemment :  
* puisque l'écriture en base 2 de $n=8$ est $(1000)_2$ et comporte 4 chiffres dont 1 chiffre égal à $1$, le nombre de multiplications est égal à $4+1=5$ multiplications ; 
* puisque l'écriture en base 2 de $n=13$ est $(1101)_2$ et comporte 4 chiffres dont 3 chiffres égaux à $1$, le nombre de multiplications est égal à $4+3=7$ multiplications ;


Ces deux nombres de multiplications excèdent de 2 les nombres de multiplications déterminé à la question **Q1** car dans l'implémentation itérative on a compté :
* pour $n=8$ : à la dernière itération, la multiplication de `v_aux` égale à 1 par `x_aux` égale à $x^{2^3}$ (non comptée en **Q1**) et la multiplication de `x_aux` par elle-même (la dernière actualisation de la variable `x_aux`) qui lui donne la valeur $x^{2^4}$, non calculée en **Q1** car non utilisée ;
* pour $n=13$ : à la première itération, la multiplication de `v_aux` égale à 1 par `x_aux` égale à $x$ (non comptée en **Q1**), et, la dernière actualisation de la variable `x_aux` qui lui donne la valeur $x^{2^4}$, non calculée en **Q1** car non utilisée).

<u>**En conclusion**</u> : $\;$ l'algorithme d'exponentiation rapide permet de réduire le nombre de multiplication nécessaire au calcul de $x^n$ d'un nombre de l'ordre de $n$ à un nombre de l'ordre de $\lfloor \log_2 n\rfloor$ dans les meilleurs cas et de l'ordre de $2.\lfloor \log_2 n\rfloor$ dans les pires cas.
