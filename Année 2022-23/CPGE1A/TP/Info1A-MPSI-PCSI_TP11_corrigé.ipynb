On adapte la fonction à une recherche de minimum, en changeant les noms `vmax` et `posmax` en `vmin` et `posmin`, le test `L[i] > vmax` en `L[i] < vmin`.   
Pour que la position du minimum soit actualisée si l'on rencontre à un rang `i` d'autres occurrences du minimum déjà trouvé lors du parcours de `L[0:i]`, on change le test `L[i] < vmin` en `L[i] <= vmin`.

```python
def posmind(L):
    vmin, posmin = L[0], 0
    for i in range(1, len(L)):
        if L[i] <= vmin:            ## <--- L[i] <= vmin assure que la position du minimum sera mise à jour
                                    ##      si une valeur inférieure strictement au minimum déjà trouvé est rencontrée
                                    ##      (cas où L[i] < vmin)
                                    ##      OU SI une valeur égale au minimum déjà trouvé est rencontrée 
                                    ##      (cas où L[i] == vmin : dans ce cas le minimum est présent plusieurs fois)
            vmin, posmin = L[i], i
    return posmin
```

**Q4.** Donner, pour l'algorithme implémenté par la fonction `posmind`, un invariant de boucle, permettant de vérifier la correction de l'algorithme.

```python
def posmind(L):
    vmin, posmin = L[0], 0
    for i in range(1, len(L)):
        if L[i] <= vmin:
            vmin, posmin = L[i], i
    return posmin
```

```python
def tri_insertion(L):
    for i in range(1, len(L)):
        # inserer L[i] dans L[0:i] déjà triée
        a_placer = L[i]
        j = i
        while j > 0 and a_placer < L[j - 1]:
            ## décaler L[j - 1] d'un rang vers la droite
            L[j] = L[j - 1]
            ## actualiser j pour l'itération suivante
            j = j - 1
            
        L[j] = a_placer
            
            
        
        
        
        
        
        
```

***Q4. corrigé***

L'invariant de boucle est, pour tout $i\in[\![0; {\rm len}(L) - 1]\!]$, la propriété, $\left({\mathcal P}_i\right)$, suivante :  
$$
\left({\mathcal P}_i\right) \; :\quad
\left(
\begin{array}{ll}
 & \mbox{À la fin de l'itération d'indice $i$,} \\
 & \mbox{$vmin$ est égal au minimum de $L[0:i+1]$} \\
 & \mbox{et $posmin$ est la position de la dernière occurrence de $vmin$ au sein de $L[0:i+1]$} \\
\end{array}
\right)
$$

On peut vérifier que la propriété est vraie pour tout $i\in[\![0; {\rm len}(L) - 1]\!]$ :  
* [initialisation] pour $i=0$, *i.e.* ***avant toute itération***, $\left({\mathcal P}_0\right)$ est vraie, car :  
`vmin` vaut $L[0]$, et `posmin`, égale à zéro, sont bien, respectivement, le minimum de $L[0:i+1]=L[0:0+1]=L[0:1]=[L[0]]$, et sa position au sein de $[L[0]]$ (puisque l'on a posé : `vmin, posmin = L[0], 0`) ;
* [hérédité] si pour un pour certain $i\in[\![1; {\rm len}(L) - 1]\!]$, $\left({\mathcal P}_{i-1}\right)$ est vraie, alors, au début de l'itération suivante, d'indice $(i-1)+1=i$ :  
on a $vmin$ qui est égal au minimum de $L[0:i]$ et $posmin$ qui est égal à la position de la dernière occurrence de $vmin$ au sein de $L[0:i+1]$.  
Si $L[i]>vmin$, alors le minimum de $L[0:i+1]$ est le même que celui de $L[0:i]$ et la position de sa dernière occurrence au sein de $L[0:i+1]$ est inchangée, égale à $posmin$, ce qui correspond bien, considérant le code de la fonction, à l'état des variables `vmin` et `posmin` à la fin de l'itération d'indice $i$, puisque dans ce cas, leur valeur est inchangée.  
Si $L[i]=vmin$, le minimum de $L[0:i+1]$ est le même que celui de $L[0:i]$ et la position de sa dernière occurrence au sein de $L[0:i+1]$ est maintenant égale à $i$, ce qui correspond bien, considérant le code de la fonction, à l'état des variables `vmin` et `posmin` à la fin de l'itération d'indice $i$, puisque dans ce cas, la  valeur de `vmin` est actualisée à la valeur $L[i]$, égale à la valeur précédente de $vmin$ et que la valeur de `posmin` est actualisée à la valeur de $i$.  
Et enfin, si $L[i]<vmin$, alors le minimum de $L[0:i+1]$ est maintenant égal à $L[i]$ et n'a jamais été atteint précédemment, car $L[i]$ est inférieur strictement au minimum de $L[0:i]$, la position de sa dernière occurrence au sein de $L[0:i+1]$ est $i$, ce qui correspond bien, considérant le code de la fonction, à l'état des variables `vmin` et `posmin` à la fin de l'itération d'indice $i$, puisque dans ce cas, la  valeur de `vmin` est actualisée à la valeur de $L[i]$ et la valeur de `posmin` est actualisée à la valeur de $i$.  
Ainsi, dans les trois cas de figure possibles, la propriété $\left({\mathcal P}_{i}\right)$, vraie à la fin de l'itération d'indice $i-1$, reste vraie à la fin de l'itération d'indice $i$.
  
On en conclut, qu'en vertu du principe de récurrence, initialisation et hérédité ayant été prouvées, que :   
<div style ="text-align:center"><b><i>
la propriété $\left({\mathcal P}_{i}\right)$ est vraie pour tout $i\in[\![0; {\rm len}(L) - 1]\!]$.<i/></b>
</div>

Le fait que $\left({\mathcal P}_{i}\right)$ soit vraie pour $i={\rm len}(L)-1$, ***i.e.*** **à la fin de la dernière itération**, établit que les valeurs en sortie de `vmin` et `posmin` sont correctes.

Le tri par insertion consiste, à chaque étape du tri d'une liste $L$, à insérer la première valeur de la partie non triée de la liste, à sa bonne place, au sein de la partie déjà triée de $L$.

***Q2. corrigé***

```python
def tri_insertion(L):
    for i in range(1, len(L)):
        # insertion de L[i] dans L[0:i] déjà triée
        a_placer = L[i]
        j = i
        while 0 < j and L[j - 1] > a_placer:
            L[j] = L[j - 1]
            j = j - 1
        L[j] = a_placer
        
## tests
from random import randint
L = [randint(1, 10) for k in range(30)]
print(L)
tri_insertion(L)
print(L)
```
