On considère ici des arbres binaires dont les noeuds sont étiquetés par des valeurs qui peuvent être comparées entre elles (le type des valeurs doit être muni d'un ordre : `int`, `float`, `str`, *etc.*).  

### <span style="color:black"> I) Généralités</span>

 **I.1. Définition**

Un arbre binaire de recherche (on abrègera en **ABR**) est un arbre qui est :  

* soit vide ;  
* soit tel que l'étiquette de sa racine est :
    - supérieure ou égale à toutes les étiquettes des noeuds de son sous-arbre gauche ;  
    - inférieure ou égale à toutes les étiquettes des noeuds de son sous-arbre droit.  
    
**et <span style="color:red">la propriété doit <u>aussi</u> être vraie pour les sous-arbres gauche et droit de l'arbre**</span>.

* exemple 1 :

```python
def creer_arbre(etiquette=None):
    if etiquette is None:
        return None
    return Noeud(None, etiquette, None)
et = creer_arbre
```

```python
ex1 = Noeud(Noeud(None, 1, et(2)), 3, et(4))
ex1
```

```python
aff1(ex1)
```

```python
ex3 = Noeud(Noeud(None, 2, Noeud(None, 1, None)), 3, Noeud(None, 4, None))
aff1(ex3)
```

 **I.2. fonctions sur les arbres toujours valables pour les arbres binaires de recherche**

```python
def taille(a):
    """renvoie le nombre de noeuds de l'arbre a"""
    if a is None:
        return 0
    else:
        return 1 + taille(a.gauche) + taille(a.droit)
```

```python
taille(ex1)
```

```python
def hauteur(a):
    """renvoie la hauteur de l'arbre a"""
    if a is None:
        return -1
    else:
        return 1 + max(hauteur(a.gauche), hauteur(a.droit))
```

```python
hauteur(ex1)
```

```python
def parcours_infixe_affichage(a):
    """affiche les étiquettes des noeuds de l'arbre a dans un parcours infixe"""
    if a is None:
        return
    parcours_infixe_affichage(a.gauche)
    print(a.valeur, end = ' ')
    parcours_infixe_affichage(a.droit)
```

```python
parcours_infixe_affichage(ex1)
```

### III) Recherche dans un ABR

 **III.1. Fonction de recherche dans un ABR**

```python
def rechABR(a, x):
    if a == None:
        return None
    ## cas général
    if a.valeur == x:
        return True
    if x < a.valeur: ## strict convient car si == on a déjà renvoyé True
        return rechABR(a.gauche, x)
    else:
        return rechABR(a.droite, x)
```

ou bien, par exemple :

```python
def appartient(x, a):
    """renvoie True ou False, selon que la valeur x est stockée ou non dans l'ABR a"""
    if a is None :
        return False
    if x < a.valeur :
        return appartient(x , a.gauche)
    elif x > a.valeur:
        return appartient(x, a.droit)
    else :
        return True
```

 **III.2. Efficacité de la fonction de recherche dans un ABR**

Quelle est l'efficacité de la fonction appartient, en terme du nombre
d'opérations effectuées pour rechercher une valeur $x$ dans un arbre $a$ qui
contient $N$ éléments ?  
* Si les éléments sont répartis à peu près équitablement
entre les deux sous-arbres,  
la fonction `appartient` élimine la moitié des éléments
à chaque étape.  

Cela fait écho à la recherche dichotomique dans un
tableau trié, où chaque étape divise par deux le nombre d'éléments à examiner
\[NSI 1ère, chap. 11\]. Comme on l'a vu, c'est extrêmement efficace. Ainsi,
il ne faut pas plus de 20 étapes pour chercher parmi un million de valeurs.
D'autre part, il n'y aucun risque de provoquer l'erreur `RecursionError` dans
ce cas.  
Cela étant, ***rien ne garantit que les éléments sont équitablement répartis***,
à chaque étape, entre les deux sous-arbres.  
  
**En particulier**, l'arbre a peut tout à fait être 
un peigne, comme l'arbre ci-dessous, ou plus généralement un arbre complètement
filiforme dont la hauteur est égale au nombre
de noeuds.   
Dans ce cas, la recherche est au contraire peu efficace, car elle
est susceptible de parcourir l'arbre tout entier, par exemple dans le cas
d'une recherche infructueuse.  
À cet égard, un ABR linéaire n'est pas meilleur
qu'une liste chârnée.  

<u>D'une manière générale</u>, **le coût de la recherche dans un ABR est majoré
par sa hauteur**.  
En effet, chaque étape de la recherche descend dans l'un
des deux sous-arbres.  
Du coup, on ne peut répéter cela un nombre de fois
plus grand que la hauteur de l'arbre, par définition.  
***La hauteur d'un ABR
dépend directement de la façon dont il a été construit***, ce qui est discuté
dans les deux sections suivantes.

### IV) Ajout dans un ABR

 **IV.1. Fonction d'ajout dans un ABR**

```python
def ajoute(x, a):
    """ajoute une valeur x dans l'ABR a"""
    if a is None :
        return Noeud(None, x, None)
    if x < a.valeur :
        return Noeud(ajoute(x, a.gauche), a.valeur, a.droit)
    else :
        return Noeud(a.gauche, a.valeur, ajoute(x, a.droit))
```

```python
ex4 = ajoute(5, ex1)
aff(ex4)
```

```python
ex5 = ajoute(1, ex1)
aff(ex5)
```

```python
ex6 = None
for i in range(1, 5):
    ex6 = ajoute(i, ex6)
aff(ex6)
```

 **IV.2. Efficacité de la fonction d'ajout dans un ABR**

**La complexité de notre fonction `ajoute` n'est pas différente de celle de
la fonction `appartient`**.   
En effet, on prend exactement les mêmes décisions
quant à la descente dans l'arbre et, comme pour la fonction `appartient`, on
fait un <u>nombre borné d'opérations à chaque étape</u>. La conclusion est donc
la même : ***la complexité dépend de la forme de l'arbre*** et elle est, **dans le
pire des cas, majorée par sa hauteur**.  
En ce qui concerne l'<u>utilisation de la
mémoire</u>, la fonction `ajoute` construit autant de noeuds qu'elle fait d'appels
récursifs. Sa **<u>complexité en espace</u>** est donc **comparable à sa complexité en
temps**.
