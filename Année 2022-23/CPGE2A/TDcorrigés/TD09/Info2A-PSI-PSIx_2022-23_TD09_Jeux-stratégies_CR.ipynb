On reprend l'algorithme usuel de recherche d'une valeur dans une liste :

```python
def appartientL(e, A):
    for etat in A:
        if etat == e:
            return True
    return False
```

On rappelle que la complexité de cet algorithme, en fonction du nombre $n$ d'éléments de $A$, est en $O(1)$ dans le meilleur cas (lorsque la valeur cherchée est en première position dans la liste), et en $O(n)$ dans le pire cas (lorsque la valeur cherchée n'est pas dans la liste).

**Q7.**

Si les attracteurs sont définis par des dictionnaires dont les numéros des états appartenant à l'attracteur sont les clés, alors on peut tester l'appartenance d'un état à l'attracteur par un test `e in A` qui est en $O(1)$, c'est-à-dire dont la complexité est indépendante du nombre d'éléments (couples (clé, valeur)) dans le dictionnaire, ce qui est particulièrement intéressant lorsque le nombre d'états du jeu est important (ce qui est le cas pour la plupart des jeux, même très simples : environ 5000 états pour le jeu de morpion étudié dans la suite).

Si les attracteurs sont donnés sous la forme de dictionnaires comme indiqué, la fonction `appartientD` sera implémenté comme suit :

```python
def appartientD(e, A):
    return e in A
```

<i><u>Note</u> :</i> cette implémentation conviendrait aussi si $A$ est de type `list` mais sa complexité, avec une entrée de type `list` ne serait pas meilleure que celle de la fonction `appartientL`.

**Q8.**

La première fonction examine tous les successeurs possibles de $e$ et renvoie `True` dès qu'un successeur élément de $A$ est trouvé, et `False` si aucun successeur n'est dans $A$ :

```python
def exist_succ(jeu, e, A, appartient):
    succs_e = jeu['transitions'][e]
    for etat in succs_e:
        if appartient(etat, A):
            return True
    return False
```

La seconde fonction teste si $e$ admet au moins un successeur, et renvoie `False` si ce n'est pas le cas. Si $e$ admet au moins un successeur, la fonction examine tous les successeurs possibles de $e$ et renvoie `False` dès qu'un successeur n'est pas élément de $A$, et `True` s'il s'avère que tous les successeurs sont dans $A$ :

```python
def forall_succ(jeu, e, A, appartient):
    succs_e = jeu['transitions'][e]
    if len(succs_e) == 0:
        return False
    for etat in succs_e:
        if not appartient(etat, A):
            return False
    return True
```

Lors d'un appel à `step`, on parcourt l'ensemble des états possibles du jeu, et pour chacun d'eux qui n'est pas déjà dans l'attracteur :
- si l'état est contrôlé par le joueur $j$, on cherche s'il existe un coup lui permettant d'atteindre un état présent dans l'attracteur ;  
- si l'état est contrôlé par l'adversaire du joueur $j$, on cherche si, quelque soit le coup choisi par l'adversaire, le jeu arrive dans un état présent dans l'attracteur.  
On termine en ajoutant à l'attracteur les états vérifiant ces conditions, et en renvoyant un booléen indiquant si de nouveaux états ont été ajoutés à l'attracteur.

```python
def step(jeu, j, A, appartient):
    ## recherche d'états à ajouter à l'attracteur
    nouveaux_etats = []
    for etat in jeu['etats']:
        if not appartientD(etat, A):
            if jeu['controles'][etat] == j:
                if exist_succ(jeu, etat, A, appartientD):
                    nouveaux_etats.append(etat)
                    flag = True
            else:
                if forall_succ(jeu, etat, A, appartientD):
                    nouveaux_etats.append(etat)
                    flag = True
    ## ajouts des nouveaux états à l'attracteur
    for etat in nouveaux_etats:
        A[etat] = None
    ## renvoie du booléen indiquant si de nouveaux états ont été trouvés
    return len(nouveaux_etats) > 0
```

**Q16.**

La fonction `controle(conf)` vérifie que le nombre de cellules occupées par le joueur 1, $n_1$, ou le joueur 2, $n_2$, sont soit égaux (c'est alors au joueur 1 et jouer), ou que $n_1=n_2+1$ (c'est alors au joueur 2 de jouer).

```python
def controle(conf):
    cpt1, cpt2 = 0, 0
    for i in range(3):
        for j in range(3):
            if conf[i][j] == symboles[1]:
                cpt1 += 1
            elif conf[i][j] == symboles[2]:
                cpt2 += 1
    assert 0 <= cpt1 - cpt2 and cpt1 - cpt2 <= 1
    if cpt1 - cpt2 == 0:
        return 1
    else:
        return 2
```

En utilisant des compréhensions de listes :

```python
def gagnante(conf, k):
    G = {1: [symboles[1]] * 3, 2: [symboles[2]] * 3} # G[1] = ['O', 'O', 'O'], G[2] = ['X', 'X', 'X'] ici
    ## alignements horizontaux
    for i in range(3):
        if [conf[i][j] for j in range(3)] == G[k]:# examen de la liste des symboles sur la ligne i
            return True
    ## alignements verticaux
    for j in range(3):
        if [conf[i][j] for i in range(3)] == G[k]:# examen de la liste des symboles sur la colonne j
            return True
    ## alignements diagonaux
    if [conf[i][i] for i in range(3)] == G[k]:# examen de la liste des symboles sur la la diagonale descendante
        return True
    if [conf[i][2 - i] for i in range(3)] == G[k]:# examen de la liste des symboles sur la la diagonale ascendante
        return True
    ## si aucun alignement de trois symbole n'a été trouvé
    return False
```

```python
def verif2(conf):
    G = {1: [symboles[1]] * 3, 2: [symboles[2]] * 3} # G = {1: ['O', 'O', 'O'], 2: ['X', 'X', 'X']} ici
    joueur = controle(conf) ## joueur contrôlant la configuration
    ## alignements horizontaux
    for i in range(3):
        if [conf[i][j] for j in range(3)] == G[joueur]:# examen de la liste des symboles sur la ligne i
            return False ## rejet de la configuration si le joueur ayant le contrôle a un alignement gagnant
    ## alignements verticaux
    for j in range(3):
        if [conf[i][j] for i in range(3)] == G[joueur]:# examen de la liste des symboles sur la colonne j
            return False ## rejet de la configuration si le joueur ayant le contrôle a un alignement gagnant
    ## alignements diagonaux
    if [conf[i][i] for i in range(3)] == G[joueur]:# examen de la liste des symboles sur la la diagonale descendante
        return False ## rejet de la configuration si le joueur ayant le contrôle a un alignement gagnant
    if [conf[i][2 - i] for i in range(3)] == G[joueur]:# examen de la liste des symboles sur la la diagonale ascendante
        return False ## rejet de la configuration si le joueur ayant le contrôle a un alignement gagnant
    return True 
```

```python
def transition(conf1, conf2):
    joueur1 = controle(conf1) 
    cpt = 0 # compteur du nombre de cases de contenu différent
    for i in range(3):
        for j in range(3):
            if conf1[i][j] != conf2[i][j]:
                if cpt == 1: ## si on a déjà trouvé une case de contenu différent
                    return False
                ## s'il s'agit de la première case de contenu différent trouvé
                if conf1[i][j] != '_' or conf2[i][j] != symboles[joueur1]:
                    return False
                cpt = 1
    # si le parcours simultané des deux tableaux termine sans que False ait été renvoyé, 
    # il reste à vérifier qu'il y a bien une case de contenu différent et non aucune
    return cpt == 1
```

**Q27.**

On propose ci-dessous une implémentation de l'algorithme décrit dans l'énoncé, tout d'abord sans utiliser de dictionnaire des configurations, puis en construisant un dictionnaire, `Dconfs2` des configurations, ce qui permet de rechercher avec une complexité constante si une configuration a déjà été obtenue, et permet d'obtenir une représentation du jeu en diminuant le temps d'exécution d'un facteur 10. 

* sans dictionnaire des configurations

```python
from time import time

t0 = time()

from collections import deque
from copy import deepcopy

## initialisations
confs2 = [plateau_vide()] ## liste des configurations possibles
conf_init = 0             ## numéro de la configuration initiale
LS2 = [0]                 ## liste des numéros d’états (position de la configuration dans confs2)
dC2 = {0: 1}              ## dictionnaire indiquant pour chaque état le joueur qui le contrôle
dAdj2 = {0: []}           ## dictionnaire donnant les états atteignables depuis chaque état
LObj12, LObj22 = [], []   ## listes des états objectifs pour les joueurs 1 et 2

## initialisation de la file
file = deque()
file.appendleft(0)
## parcours en largeur du graphe des états du jeu
while len(file) > 0:
    # on défile un numéro d'état (ce numéro donne la position de la configuration correspondante dans la liste confs2)
    conf_no = file.pop()
    # récupération de la configuration correspondante
    conf = confs2[conf_no]
    # récupération du numéro du joueur contrôlant la configuration
    joueur = dC2[conf_no]
    ## pour chaque case vide dans la configuration défilée, on examine toutes les configurations possibles au coup suivant 
    ## pour cela on parcourt la grille à l'aide de deux boucles imbriquées
    for i in range(3):
        for j in range(3):
            ## si une case est vide :
            if lire(conf, i, j) == 0:
                ## on en crée une copie profonde
                next_conf = deepcopy(conf) ## ou bien next_conf =  [[conf[i][j] for j in range(3)] for i in range(3)]
                ## on fait occuper la case vide par le joueur contrôlant la confinguration
                jouer(next_conf, i, j, joueur)
                ## on doit maintenant examiner si la configuration construite, next_conf, a déjà été trouvée ou non :
                ##  - si elle a déjà été atteinte, on doit récupérer le numéro d'état correspondant, i.e. sa position
                ##    dans la liste confs2 ;
                ##  - si elle n'a jamais été atteinte, on doit l'ajouter à la liste confs2, et actualiser la liste
                ##    des numéros d'états, ainsi que le dictionnaire dC2, et, si elle est gagnante, l'ajouter à LObj12
                ##    ou LObj22.
                
                ## On initialise le numéro de la configuration à la valeur len(confs2), qui sera sa position dans 
                ## la liste confs2 si on doit l'y ajouter
                next_conf_no = len(confs2)
                ## puis on parcourt la liste confs2 : si on trouve la configuration, on donne à next_conf_no
                ## la valeur de la position de la configuration dans la liste confs2
                for k in range(len(confs2)):
                    if next_conf == confs2[k]:
                        next_conf_no = k ## <- si la configuration n'est pas nouvelle,
                                         ## son numéro est donné par sa position dans la liste confs2
                        break ## on termine la boucle si la configuration a été trouvée 
                ## on examine si la configuration a été trouvée en examinant si next_conf_no a toujours pour valeur
                ## len(confs2) ou non, si tel est le cas, il s'agit d'une nouvelle configuration et on doit l'ajouter
                ## à la fin de la liste confs2, et procéder aux actualisations nécessaires
                if next_conf_no == len(confs2):
                    confs2.append(next_conf)
                    ## on ajoute le numéro de cette nouvelle configuration à la liste des numéros d'état
                    LS2.append(next_conf_no)
                    ## on compléte le dictionnaire dC2 pour indiquer quel joueur contrôle cet état
                    dC2[next_conf_no] = 3 - joueur ## 3 - joueur calcul le numéro du joueur adversaire
                    ## on crée une nouvelle clé dans le dictionnaire dAdj2 pour ce nouvel état, associée à []
                    dAdj2[next_conf_no] = []
                    ## on examine si la configuration est gagnante pour l'un ou l'autre joueur, et si tel est le cas,
                    ## on l'ajoute à la liste des objectifs du joueur
                    if gagnante(next_conf, joueur):
                        if joueur == 1:
                            LObj12.append(next_conf_no)
                        else:
                            LObj22.append(next_conf_no)
                    ## si la configuration nouvelle n'est pas gagnante, on l'ajoute à la file des configurations
                    ## à examiner afin de déterminer ses successeurs possibles
                    if not gagnante(next_conf, joueur):
                        file.appendleft(next_conf_no)
                ## dans tous les cas (que la configuration soit nouvelle ou non), on doit l'ajouter en tant que
                ## successeur possible de la configuration conf
                dAdj2[conf_no].append(next_conf_no)

t = time() - t0
print(t) ## exemple de temps d'exécution : 9.401000022888184

print(len(confs2), len(LS2), len(dC2), len(dAdj2), len(LObj12), len(LObj22)) ## 5478 5478 5478 5478 626 316
print(len(dAdj2), sum([len(dAdj2[etat]) for etat in dAdj2.keys()])) ## ordre et taille du graphe : (5478, 16167)
```
