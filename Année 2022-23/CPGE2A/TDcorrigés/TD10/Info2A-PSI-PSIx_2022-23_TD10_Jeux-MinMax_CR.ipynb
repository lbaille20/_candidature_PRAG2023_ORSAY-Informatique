**Q6.** Définir une fonction `jouer_colonne(jeu, joueur, col)`, qui joue un coup dans la colonne `col` pour le joueur `joueur`.  
La fonction agira par effet, en modifiant l'état du jeu, et renverra les coordonnées de la case jouée.  
On inscrira, sur plateau de jeu, le numéro du joueur (1 ou 2) à la place occupée par le jeton joué et on actualisera le total du nombre de coups joués et les nombres de jetons dans les différentes colonnes.  
On vérifiera, en vérifiant des assertions, que le joueur est bien le prochain joueur devant jouer, et que la colonne jouée n'est pas remplie.

```python
def jouer_colonne(jeu, joueur, col):
    assert joueur == prochain_joueur(jeu)
    assert col in colonnes_jouables(jeu)
    hcol = jeu['hauteurs'][col]
    row = nrows - 1 - hcol
    jeu['plateau'][row][col] = joueur
    jeu['hauteurs'][col] += 1
    jeu['nbcoups'] += 1
    return row, col
```

**Q7.** Écrire de même une fonction `dejouer_colonne(jeu, col)` rétablissant, après exécution d'un appel `jouer_colonne(jeu, joueur, col)`, le jeu dans son état précédent.

```python
def dejouer_colonne(jeu, col):
    hcol = jeu['hauteurs'][col] - 1
    row = nrows - 1 - hcol
    jeu['plateau'][row][col] = 0
    jeu['hauteurs'][col] -= 1
    jeu['nbcoups'] -= 1
```

**Q8.** Construire la liste, `Lalignements`, de tous les alignements de quatre cases possibles sur le plateau de jeu, sous la forme d'une liste de tuples représentant les coordonnées, $(i,j)$, de chaque case de l'alignement.  

* par exemple, on trouvera dans la liste `Lalignements`, les listes $[(0, 0), (0, 1), (0, 2), (0, 3)]$ et $[(3, 1), (2, 2), (1, 3), (0, 4)]$ correspondant aux alignements suivants :  
<img src="https://github.com/lbaille20/Info2A_2022-23/raw/main/Info2A-PSIx/aligns-1-2.png"/>

On pourra utiliser la liste `vecteurs_deplacements` des coordonnées des quatre vecteurs de déplacements, $u_1, u_2, u_3, u_4$ de coordonnées respectives $(0, 1)$ (déplacement horizontal), $(1, 0)$ (déplacement vertical), $(1, 1)$ (déplacement descendant) et $(-1, 1)$ (déplacement ascendant) :

```python
vecteurs_deplacements = [(0, 1), (1, 0), (1, 1), (-1, 1)]
```

et la fonction suivante :

```python
def translation(M: tuple, u: tuple, k: int) -> tuple:
    """renvoie les coordonnées de l'image du point M par la translation de vecteur k * u"""
    i, j = M
    ui, uj = u
    return (i + k * ui, j + k * uj)
```

afin de générer les alignements à l'aide de compréhensions de listes.

Ainsi, les alignements donnés en exemple ci-dessus pourront être générés, respectivement, par :

```python
[translation((0, 0), vecteurs_deplacements[0], k) for k in range(4)]
```

```python
[translation((3, 1), vecteurs_deplacements[3], k) for k in range(4)]
```

**Construction de la liste `Lalignements` :**

```python
Lalignements = [] #24 = 4 * 6 horizontaux ; 21 = 3 * 7 verticaux ; 12 = 3 * 4 descendants ; 12 = 3 * 4 ascendants : total 69
## alignements horizontaux
cpt1 = 0
for i in range(nrows):
    for j in range(ncols - 3):
        Lalignements.append([translation((i, j), vecteurs_deplacements[0], k) for k in range(4)])
        cpt1 += 1
print("alignements horizontaux :", cpt1)
## alignements verticaux
cpt2 = 0
for j in range(ncols):
    for i in range(nrows - 3):
        Lalignements.append([translation((i, j), vecteurs_deplacements[1], k) for k in range(4)])
        cpt2 += 1
print("alignements verticaux :  ", cpt2)
## alignements descendants
cpt3 = 0
for i in range(nrows - 3):
    for j in range(ncols - 3):
        Lalignements.append([translation((i, j), vecteurs_deplacements[2], k) for k in range(4)])
        cpt3 += 1
print("alignements descendants :", cpt3)
## alignements ascendants
cpt4 = 0
for i in range(3, nrows):
    for j in range(ncols - 3):
        Lalignements.append([translation((i, j), vecteurs_deplacements[3], k) for k in range(4)])
        cpt4 += 1
print("alignements ascendants : ", cpt4)
```

**Q9.** Construire le dictionnaire `Dalignements`, dont les clés sont les couples d'entiers des coordonnées des cases du plateau, avec, pour valeur associée à chaque clé $(i, j)$, la liste des alignements de quatre cases passant par la case de coordonnées $(i, j)$.  
On pourra recourir à des tests `x in L`, testant si une valeur $x$ est dans une liste $L$, de complexité $O({\rm len}(L))$ dans le pire cas.  
Par exemple, pour la case en haut à gauche, `Dalignements[(0, 0)]`, devra renvoyer à la liste de trois alignements `[[(0, 0), (0, 1), (0, 2), (0, 3)], [(0, 0), (1, 0), (2, 0), (3, 0)], [(0, 0), (1, 1), (2, 2), (3, 3)]]`.

```python
Dalignements = {}
# initialisation
for i in range(nrows):
    for j in range(ncols):
        Dalignements[(i, j)] = []
# calcul
for alignement in Lalignements:
    for coords in alignement:
        Dalignements[(coords[0], coords[1])].append(alignement)
```

**Q10.** Définir une fonction `evaluer(jeu, alignement)`, renvoyant le numéro du joueur occupant les quatre cases de l’alignement si tel est le cas, et zéro sinon.

```python
def evaluer_alignement(jeu, alignement):
    L = [0] * 3
    for coords in alignement:
        L[jeu['plateau'][coords[0]][coords[1]]] += 1
    for joueur in (1, 2):
        if L[joueur] == 4:
            return joueur
    return 0
```

**Q15.** Définir les fonctions `maximin(jeu, n)` et `minimax(jeu, n)`. Pour $n>0$, on créera, pour chaque colonne jouable, une copie du jeu dans laquelle on jouera la colonne, et on sélectionnera parmi ces colonnes, celle qui amène à l’horizon $n-1$, l’heuristique – calculée avec la fonction de l’adversaire, de plus grande valeur si c’est au joueur 1 de jouer, celle de plus petite valeur si c’est au joueur 2 de jouer : la fonction renverra ce numéro de colonne, ainsi que la valeur de l’heuristique qu’il est possible d’atteindre. Si une colonne est gagnante pour le joueur, on interrompra la recherche et on renverra le numéro de la colonne et la valeur infinie de l’heuristique correspondante.  
Pour $n=0$ ou si le plateau est rempli, les deux fonctions renverront la valeur `None` et l’heuristique du jeu.

```python
def maximin(jeu, n):
    colonnes_possibles = colonnes_jouables(jeu)
    ## cas terminaux
    if n == 0 or len(colonnes_possibles) == 0:
        return None, heuristique(jeu)
    ## cas général
    colmax, hmax = None, -float('inf')
    for col in colonnes_possibles:
        jeu_suivant = copie(jeu)
        coords = jouer_colonne(jeu_suivant, 1, col)
        if teste_gagnante(jeu_suivant, 1, coords):
            return col, float('inf')
        eval_minimax = minimax(jeu_suivant, n - 1)[1]
        if eval_minimax > hmax:
            colmax, hmax = col, eval_minimax
    return colmax, hmax
```

* fonction `minimax(jeu, n)` :

```python
def minimax(jeu, n):
    colonnes_possibles = colonnes_jouables(jeu)
    ## cas terminaux
    if n == 0 or len(colonnes_possibles) == 0:
        return None, heuristique(jeu)
    ## cas général
    colmin, hmin = None, float('inf')
    for col in colonnes_possibles:
        jeu_suivant = copie(jeu)
        coords = jouer_colonne(jeu_suivant, 2, col)
        if teste_gagnante(jeu_suivant, 2, coords):
            return col, -float('inf')
        eval_maximin = maximin(jeu_suivant, n - 1)[1]
        if eval_maximin < hmin:
            colmin, hmin = col, eval_maximin
    return colmin, hmin
```

**Q17.** Proposer une implémentation pour les deux fonctions modifiées, que l’on nommera `maximin_alpha_beta(jeu, alpha, beta, n)` et `minimax_alpha_beta(jeu, alpha, beta, n)`.

```python
def maximin_alpha_beta(jeu, alpha, beta, n):
    colonnes_possibles = colonnes_jouables(jeu)
    if n == 0 or len(colonnes_possibles) == 0:
        return None, heuristique(jeu)
    ## recherche de l'heuristique maximale à l'horizon de n coups
    colmax, hmax  = None, -float('inf')
    for col in colonnes_possibles:
        jeu_suivant = copie(jeu)
        coords = jouer_colonne(jeu_suivant, 1, col)
        if teste_gagnante(jeu_suivant, 1, coords):
            return col, float('inf')
        eval_minimax = minimax_alpha_beta(jeu_suivant, alpha, beta, n - 1)[1]
        if eval_minimax >= beta:      # la valeur minimale sera inférieure à alpha 
            return col, eval_minimax  # donc ne sera pas retenue par maximin(..., n + 1)
        if eval_minimax > hmax:
            colmax, hmax = col, eval_minimax
        alpha = max(alpha, hmax) ## mise à jour de la valeur de alpha en vue d'appels ultérieurs à minimax
    return colmax, hmax
```

```python
def minimax_alpha_beta(jeu, alpha, beta, n):
    colonnes_possibles = colonnes_jouables(jeu)
    if n == 0 or len(colonnes_possibles) == 0:
        return None, heuristique(jeu)
    colmin, hmin  = None, float('inf')
    for col in colonnes_possibles:
        jeu_suivant = copie(jeu)
        coords = jouer_colonne(jeu_suivant, 2, col)
        if teste_gagnante(jeu_suivant, 2, coords):
            return col, -float('inf')
        eval_maximin = maximin_alpha_beta(jeu_suivant, alpha, beta, n - 1)[1]
        if eval_maximin <= alpha:     # la valeur minimale sera inférieure à alpha 
            return col, eval_maximin  # donc ne sera pas retenue par maximin(..., n + 1)
        if eval_maximin < hmin:
            colmin, hmin = col, eval_maximin
        beta = min(beta, hmin) ## mise à jour de la valeur de beta en vue d'appels ultérieurs à maximin
    return colmin, hmin
```
