### 1.3. Mise en place de la méthode des $k$-plus proches voisins ($k$-nearest neighbors, knn)

**Q4.**

Entre deux fleurs dont les coordonnées, (longueur, largeur), sont données par les couples $(x, y)$ et $(x', y')$, la distance est donnée par le calculde la distance euclidienne :  
$$\Delta =  \sqrt{\left((x'-x)^2+(y'-y)^2\right)}.$$

Si les coordonnées sont transmises à la fonction sous la forme de deux tuples, la fonction peut s'implémenter comme suit :

```python
from math import sqrt

def distance(t1, t2):
    return sqrt((t2[0] - t1[0]) ** 2 + (t2[1] - t1[1]) ** 2)
```

**Q5.**

Une fleur à classifier étant décrite par un couple $t=(longueur, largeur)$ étant donnée, on veut renvoyer une liste dont les éléments sont les couples, (distance, espece), dont chacun donne la distance de la fleur à classifier à une fleur de l'échantillon (première composante du couple) et l'espèce à laquelle appartient la fleur de l'échantillon utilisée pour le calcul de distance (deuxième composante du couple).

```python
def liste_distances(data, t):
    x0, y0 = t ## longueur, largeur pour la fleur à classer
    L = []
    for fleur in data:
        x, y, espece = fleur #longueur, largeur, espece pour cette fleur de l'échantillon
        ## ou bien :
        ## x, y = fleur[0], fleur[1]
        dist = distance((x0, y0), (x, y))
        L.append((dist, espece))
    return L
```

```python
liste_distances(data, (2.5, 1.0)) # pour une fleur dont la longueur de pétale est 2.5 et la largeur 1.0
```

**Q6.**

On rappelle une implémentation du tri-insertion en place pour une liste de valeurs comparables entre elles par `<=` :

```python
def tri_insertion(L):
    for i in range(1, len(L)):
        # insertion de la valeur L[i] (recopiée dans tmp) dans la liste déjà triée L[:i]
        tmp = L[i]
        j = i ## j désignera la position à laquelle recopier tmp
        while j > 0 and L[j - 1] > tmp:
            L[j] = L[j - 1] ## décalage d'un rang à droite de toutes les valeurs de la partie triée qui sont > tmp
            j = j - 1 # on décremente j pour examiner la valeur suivante (à gauche de la précédente)
        L[j] = tmp
```

* tests

```python
from random import randint

L = [randint(1, 50) for i in range(50)]
print(L)
tri_insertion(L)
print(L)
```

**Il s'agit ici d'adapter l'implémentation du tri-insertion en place** au fait que la liste est une liste de couples, (distance, espece), que l'on doit trier **dans l'ordre croissant des valeurs de distance**.

```python
def trier(L):
    for i in range(1, len(L)):
        tmp = L[i]
        j = i
        while j > 0 and L[j - 1][0] > tmp[0]: ## ce sont uniquement les valeurs de la première composante que l'on compare
            L[j] = L[j - 1]
            j = j - 1 
        L[j] = tmp
```

La complexité du tri est celle du tri-insertion, en $O(n^2)$ où $n={\rm len}(L)$, dans le pire cas - liste triée initialement dans l'ordre décroissant ($O(n)$ dans le meilleur cas - liste déjà triée).

### 1.4. Algorithme $knn$

**Q7.**

Il s'agit d'examiner les $k$ plus proches voisins d'une fleur à classifier (dont on connaît longueur et largeur des pétales).

Sur l'exemple précédent, si on prend $k=5$, les $k$ plus proches voisins sont donnés par :

```python
L = liste_distances(data, (2.5, 1.0)) # exemple de la question Q5.
trier(L)
L[:5]
```

L'espèce majoritaire parmi les cinq fleurs les plus proches de la fleur à classifier est l'espèce 0 (iris setosa).

Les espèces étant étiquetées par les entiers 0, 1 et 2, on peut utiliser une liste à trois éléments pour implémenter les compteurs des trois espèces, ce qui amène à l'implémentation suivante :

```python
def classification_knn(data, t, k):
    L = liste_distances(data, t)
    trier(L)
    compteurs = [0, 0, 0 ] # une liste de trois compteurs compteurs[i] comptant les fleurs de l'espèce i
    for i in range(k): # pour les k plus petites distances obtenues
        espece = L[i][1] # on note l'espèce de la fleur à cette distance
        compteurs[espece] += 1 # on incrémente de 1 le compteur correspondant
    ## recherche du maximum
    nbmax, jmax = compteurs[0], 0 # on initialise avec le nombre de fleurs de l'espèce zéro parmi les k plus proches
    for j in range(3): ## on examine les compteurs des espèces 1 et 2
        if compteurs[j] > nbmax:
            nbmax, jmax = compteurs[j], j
    return jmax #no de l'espèce majoritaire parmi les k plus proches
```

```python
classification_knn(data, (2.5, 1.0), 10) ## avec la fleur test de la question Q5. et k = 10
```

On peut aussi utiliser un dictionnaire pour implémenter les compteurs des trois espèces (ce qui permettrait d'utiliser les noms des espèces plutôt que de les numéroter à partir de zéro) :

```python
def classification_knn1(data, t, k):
    L = liste_distances(data, t)
    trier(L)
    # un dictionnaire de compteurs Dcompteurs[espece] comptant les fleurs de l'espèce espece
    Dcompteurs = {0: 0, 1: 0, 2: 0}
    for i in range(k): # pour les k plus petites distances obtenues
        espece = L[i][1] # on note l'espèce de la fleur à cette distance
        Dcompteurs[espece] += 1 # on incrémente de 1 le compteur correspondant
    ## recherche du maximum
    nbmax, emax = Dcompteurs[0], 0 ## on initialise avec le nombre de fleurs de l'une des espèces
                                   ## parmi les k plus proches
    for esp in Dcompteurs.keys(): ## on parcourt le dictionnaire des compteurs par les clés (qui sont les espèces)
        if Dcompteurs[esp] > nbmax:
            nbmax, emax = Dcompteurs[esp], esp
    return emax #espèce majoritaire parmi les k plus proches
```

**Q8.**

La complexité d'un appel à la fonction `classification_knn(data, t, k)` en fonction du nombre $n$ de fleurs dans l'échantillon, du nombre $p$ d'espèces présentes, et de $k$, est la somme des complexités suivantes :  
- $O(n)$ pour le calcul de toutes les distances ;  
- $O(n^2)$ pour le tri de la liste des couples (distance, espèce) ;  
- $O(p+k)$ pour le dénombrement des espèces parmi les $k$ plus proches ($O(p)$ pour la construction de la liste des compteurs (liste de $p$ zéros) et $O(p)$ pour l'examen de l'espèce des $k$ plus proches) ;
- $O(p)$ pour la recherche de l'espèce majoritaire parmi les $k$ plus proches.  
  
La complexité totale est donc en $O(n)+O(n^2)+O(p+k)+O(p)=O(n^2+p+k)$.  
Le nombre d'espèces étant constant, la complexité sera en $O(n^2+k)$.

**Q9.**

La complexité peut être améliorée par le fait qu'il est inutile de trier toute la liste $L$ des couples (distance, espèces).  
On peut se contenter de parcourir la liste $L$ en construisant, au fil du parcours, la liste triée $Lk$ des $k$ plus proches voisins.  
La liste $Lk$ est initialement vide, puis, au fil du parcours de $L$ avec une boucle `for` d'indice $i$ :
- si $Lk$ comporte moins de $k$ éléments, on ajoute $L[i]=(d_i, e_i)$ en dernière position dans $Lk$ ;
- si $Lk$ comporte $k$ éléments, on teste si $L[i][0]$ (distance à la fleur à classifier) est inférieure strictement à la distance $Lk[-1][0]$ et si tel est le cas, on remplace $Lk[-1]$ par $L[i]$ ;
- on déplace le dernier couple dans la liste $Lk$ jusqu'à ce que $Lk$ soit triée (selon le principe du tri insertion). 
  
On suppose (bien sûr) que $k$ est inférieur strictement au nombre d'éléments dans la base.

```python
def classification_knn2(data, t, k):
    L = liste_distances(data, t)
    Lk = []
    for i in range(len(L)):
        if len(Lk) < k:
            Lk.append(L[i])
        else:
            if L[i][0] < Lk[-1][0]:
                Lk[-1] = L[i]
        # décalage de la dernière valeur dans Lk pour maintenir Lk triée
        tmp = Lk[len(Lk) - 1]
        j = len(Lk) - 1
        while j > 0 and Lk[j - 1][0] > tmp[0]: ## ce sont uniquement les valeurs de la première composante que l'on compare
            Lk[j] = Lk[j - 1]
            j = j - 1 
        Lk[j] = tmp
    # à ce stade Lk est la liste des couples (distance, espèce) pour les k plus proches
    
    # un dictionnaire de compteurs Dcompteurs[espece] comptant les fleurs de l'espèce espece
    Dcompteurs = {0: 0, 1: 0, 2: 0}
    for i in range(k): # pour les k plus petites distances obtenues
        espece = Lk[i][1] # on note l'espèce de la fleur à cette distance
        Dcompteurs[espece] += 1 # on incrémente de 1 le compteur correspondant
    ## recherche du maximum
    nbmax, emax = Dcompteurs[0], 0 ## on initialise avec le nombre de fleurs de l'une des espèces
                                   ## parmi les k plus proches
    for esp in Dcompteurs.keys(): ## on parcourt le dictionnaire des compteurs par les clés (qui sont les espèces)
        if Dcompteurs[esp] > nbmax:
            nbmax, emax = Dcompteurs[esp], esp
    return emax #espèce majoritaire parmi les k plus proches
```

* tests

```python
classification_knn2(data, (2.5, 1.0), 3) ## avec la fleur test de la question Q5. et k = 3
```

```python
classification_knn2(data, (2.5, 1.0), 10) ## avec la fleur test de la question Q5. et k = 10
```

Avec cette modification, la complexité des opérations de tri correspond $n$ insertions dans une liste, $Lk$, comportant au plus $k$ éléments, donc avec une complexité totale (dans le pire cas) en $O(n\times k)$, au lieu de $O(n^2)$.  
La complexité totale de `classification_knn2(data, t, k)` est donc désormais en $O(n)+O(nk)+O(p+k)+O(p)=O(nk+p)$.  
Le nombre d'espèces étant constant, la complexité sera en $O(nk)$.
