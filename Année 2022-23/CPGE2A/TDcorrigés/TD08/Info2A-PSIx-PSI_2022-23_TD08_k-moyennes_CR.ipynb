```python
from random import random

def alea_bornes(a, b):
    return a + random() * (b - a)
```

La fonction permet de simuler le tirage au hasard d'un flottant dans l'intervalle $[a, b]$.

En effet, `random()` renvoie un flottant aléatoire $\lambda$ de l'intervalle $[0, 1[$, et on a $a\leqslant a +\lambda (b-a)< b$.

**Q4.**

```python
def initialise_centres(pts, k):
    p = len(pts[0])
    centres = [[0] * p for i in range(k)]
    for j in range(p):
        vmin = min([pt[j] for pt in pts])
        vmax = max([pt[j] for pt in pts])
        for i in range(k):
            centres[i][j] = alea_bornes(vmin, vmax)
    return centres
```

La fonction permet, pour tirer au hasard $k$ centres de classes, c'est-à-dire $k$ $p$-uplets $(x_1, x_2, \ldots, x_p)$, tels que pour tout $i$ compris entre 1 et $p$, $x_i$ est compris entre les valeurs $x_{i, min}$ et $x_{i, max}$ pour les éléments de l'échantillon.

Pour notre exemple, on a $p=2$ et pour chaque centre, on tire au hasard une abscisse et une ordonnée, respectivement, comprises entre le minimum et le maximum des longueurs de pétales (abscisse) et le minimum et le maximum des longueurs de pétales (ordonnée).

```python
from math import sqrt

def distance(pt1, pt2):
    return sqrt((pt1[0] - pt2[0]) ** 2 + (pt1[1] - pt2[1]) ** 2)

def distance(pt1, pt2):
    p = len(pt1)
    return sqrt(sum([(pt1[i] - pt2[i]) ** 2 for i in range(p)]))
```

**Q5.**

```python
def classifier_point(pt, centres):
    distances = []
    for c in centres:
        distances.append(distance(pt, c))
    dmin, imin = distances[0], 0
    for i in range(1, len(distances)):
        if distances[i] < dmin:
            dmin, imin = distances[i], i
    return imin
```

**Q6.**

```python
def calcul_classes(pts, centres):
    classes = [[] for j in range(len(centres))]
    for i in range(len(pts)):
        ind_espece = classifier_point(pts[i], centres)
        classes[ind_espece].append(i)
    return classes
```

On a figuré les centres par des croix de trois couleurs différentes, et coloré chaque point représentant un élément de l'échantillon avec la couleur du centre dont il est le plus proche.

```python
def calcul_centre_classe(pts, C):
    p = len(pts[0])
    centre =  [0] * p
    for j in range(p):
        centre[j] = sum(pts[i][j] for i in C) / len(C)
    return centre
```

```python
def calcul_centre_classe(pts, C):
    p = len(pts[0])
    return [sum(pts[i][j] for i in C) / len(C) for j in range(p)]
```
