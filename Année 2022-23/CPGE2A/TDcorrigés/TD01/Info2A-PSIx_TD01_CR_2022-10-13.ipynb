On peut imaginer deux solutions, que l'on discriminera suivant leur complexité.

* solution en construisant pas à pas la matrice d'adjacence

On initialise la liste de liste `M1` à une liste vide, et à l'aide d'une boucle `for` d'indice $i$, on y ajoute au fil de ses itérations, $n$ listes vides, une pour chaque ligne de la matrice $M_1$.  
On remplit chaque ligne, d'indice $i$, $0\leqslant i\leqslant n - 1$, grâce à une boucle `for` d'indice $j$, imbriquée dans la première. À chaque itération de la boucle intérieure, on ajoute un nouveau coefficient de la matrice dans la liste de liste, selon que l'arc $(i, j)$ est présent dans la liste `LArcs` ou non.

```python
LS
```

```python
LArcs
```

```python
n = len(LS) # nombre de sommets (ordre) du graphe 1
M1 = []
for i in range(n):
    M1.append([]) # ajout d'une sous-liste vide (représentera la ligne i de la matrice)
    for j in range(n):
        #i et j correspondent respectivement à un indice de ligne et de colonne dans la matrice M1
        if (i, j) in LArcs: # si l'arc (i, j) existe dans le graphe
            M1[i].append(1) # ajout d'un coefficient 1, sera accessible par l'expression M1[i][j]
        else:               # si l'arc (i, j) n'existe pas dans le graphe
            M1[i].append(0) # ajout d'un coefficient 0, sera accessible par l'expression M1[i][j]
```

Analyse de complexité :

```python
n = len(LS) # O(1)
M1 = []
for i in range(n):
    M1.append([]) # O(1) [répété n fois]
    for j in range(n):
        #i et j correspondent respectivement à un indice de ligne et de colonne dans la matrice M1
        if (i, j) in LArcs: # O(m) m : le nombre d'éléments dans LArcs (nombre d'arcs du graphe, taille du graphe)
            M1[i].append(1) # O(1)
        else:               # 
            M1[i].append(0) # O(1)
```

Évaluation de la complexité :

$$T(n, m) = O(1) + \sum_{i=0}^{n - 1} \left(\sum_{j=0}^{n - 1} O(m) + O(1)\right)=\sum_{i=0}^{n - 1} \left(\sum_{j=0}^{n - 1} O(m)\right)
= O\left(\sum_{i=0}^{n - 1} \left(\sum_{j=0}^{n - 1} m\right)\right)
= O\left(\sum_{i=0}^{n - 1} n\times m\right)
= O\left(n^2m\right)
$$

Analyse de complexité :

On se donne une construction explicite de toutes les listes afin de ne pas être piégés dans l'analyse par des éléments de complexité cachés :

```python
n = len(LS) # O(1)

M1 = [] # O(1)
for i in range(n):
    M1.append([]) # O(1) [répété n fois]
    for j in range(n):
        M1[i].append(0) # O(1) [répété n x n  fois]

for arc in LArcs:
    M1[arc[0]][arc[1]] = 1  # O(1) [répété m fois]
```

Évaluation de la complexité :

$$T(n, m)
= O(1) + \sum_{i=0}^{n - 1} \left(O(1) + \sum_{j=0}^{n - 1} O(1)\right) + O\left(m\right)
= O(1) + \sum_{i=0}^{n - 1} \left(O(1) + n \times  O(1)\right) + O\left(m\right)
= O(1) + \sum_{i=0}^{n - 1} O(n) + O\left(m\right)
$$

$$T(n, m)
= O(1) + n\times O(n) + O\left(m\right)
= O(1) + O\left(n^2\right) + O\left(m\right)
= O\left(n^2  + m\right).
$$

avec des abrégés de constructions

```python
n = len(LS) # O(1)

M1 = [[0] * n for i in range(n)] # O(n^2)

for arc in LArcs:
    M1[arc[0]][arc[1]] = 1  # O(1) [répété m fois]
```

On retrouve bien :
    $$T(n, m) = O\left(n^2\right) + O(m) = O(n^2+m).$$

**On en conclut que la seconde méthode est préférable car sa complexité est moindre : $O(n^2+m)$ au lieu de $O(n^2m)$** (car on évite de tester, pour chaque couple de sommets, s'il est un arc du graphe).

Compte-tenu des réponses aux questions précédentes, on pourra écrire dans le fichier `fonctions_TD01.py`les définitions suivantes :

```python
def lire1_graphe(r, f):
    ## concaténation du chemin vers le répertoire et ajout de l'extension au nom de fichier
    chemin_vers_fichier = r + "\\" + f + '.txt'
    
    ## lecture du fichier
    f = open(chemin_vers_fichier)
    ligne1 = f.readline()
    lignes_suivantes = f.radlines()
    f.close()
    
    ## construction de la liste des sommets
    LS = [int(val) for val in ligne1.strip().split(' ')]
    ## ordre du graphe
    n = len(LS)
    ## construction de la liste des arcs
    Laux = []
    for elt in lignes_suivantes:
        Laux.append(elt.strip().split('\t'))
    # construction de la liste LArcs
    LArcs = []
    for elt in Laux:
        a, b = elt
        LArcs.append((int(a), int(b)))
    ## taille du graphe
    m = len(LArcs)
    
    return n, m, LS, LArcs
```

On remplacera ici les paramètres $r$ et $f$ par un unique paramètre `chemin_vers_fichier`, car la fontion précédente ne convient pas dans Capytale car il faudrait remplacer le séparateur entre le chemin vers le répertoire et le nom de fichier, `"\\"`, par `"/"` (utilisé sur plate-forme UNIX et non Windows).

```python
def lire1_graphe(chemin_vers_fichier):
    ## lecture du fichier
    f = open(chemin_vers_fichier)
    ligne1 = f.readline()
    lignes_suivantes = f.readlines()
    f.close()
    
    ## construction de la liste des sommets
    LS = [int(val) for val in ligne1.strip().split(' ')]
    ## ordre du graphe
    n = len(LS)
    ## construction de la liste des arcs
    Laux = []
    for elt in lignes_suivantes:
        Laux.append(elt.strip().split('\t'))
    # construction de la liste LArcs
    LArcs = []
    for elt in Laux:
        a, b = elt
        LArcs.append((int(a), int(b)))
    ## taille du graphe
    m = len(LArcs)
    
    return n, m, LS, LArcs
```

En cohérence, on remplacera les paramètres $r$ et $f$ par un unique paramètre `chemin_vers_fichier` pour aussi la fonction `charger1_graphe`

```python
def arcs_vers_mat1(n, LArcs):
    M = [[0] * n for i in range(n)]
    for arc in LArcs:
        M[arc[0]][arc[1]] = 1
    return M
```

```python
def charger1_graphe(chemin_vers_fichier):
    n, m, LS, LArcs = lire1_graphe(chemin_vers_fichier) # <-- dépaquetage de tuple
    return arcs_vers_mat1(n, LArcs)

# ou bien
def charger1_graphe(chemin_vers_fichier):
    n, _, _, LArcs = lire1_graphe(chemin_vers_fichier) # <-- dépaquetage de tuple avec utilisation de deux variables
                                                       #     "variables anonymes", _, pour les éléments renvoyés qui
                                                       #     n'ont pas d'utilité ici
    return arcs_vers_mat1(n, LArcs)
```

Pour cette question, on doit disposer du moyen d'accéder au nom d'un sommet à partir de son numéro. 

On peut réaliser l'opération comme suit, en construisant un dictionnaire *réciproque* du dictionnaire donnant la numérotation des sommets.

Si `DS = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4}`, le dictionnaire donnant les noms en fonction des numéros, que l'on note ici `DNS` sera `{0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E'}` et on peut l'obtenir ainsi :

```python
DS = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4}
DNS = {}
for cle, valeur in DS.items():
    DNS[valeur] = cle
```

```python
DNS
```

Étant convenu que les numéros de sommets forment une suite de la forme $0, 1, \ldots, n - 1$, où $n$ est l'ordre du graphe, une liste des sommets, `LNS`, dans laquelle on place le nom de chaque sommet en fonction de son numéro, aurait convenu aussi pour réaliser l'objectif visé (à savoir associer un $1$ en position $i, j$ dans la liste de listes représentant la matrice d'adjacence au bon arc $(u, v)$ où $u$ et $v$ sont des noms de sommets.

```python
DS = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4}
LNS = [''] * len(DS)
for cle, valeur in DS.items():
    LNS[valeur] = cle
print(LNS)
```

Il serait alors possible de remplacer la définition du dictionnaire `DNS` par celle d'une liste `LNS`, et `DNS[i]` et `DSN[j]` par `LNS[i]` et `LSN[j]`, dans le code de la fonction ci-dessous, sans en affecter le comportement.

```python
def Madj_to_DLadj(DS, Madj):
    # construction du dictionnaire réciproque DNS
    DNS = {}
    for sommet, numero in DS.items():
        DNS[numero] = sommet
    # on initialise le dictionnaire DLadj
    #         avec pour clés tous les noms de sommets, et valeurs associées des listes vides
    DLadj = {}
    for s in DS.keys():
        DLadj[s] = []
    
    # on parcourt la matrice d'adjacence, représentée en machine par un tableau à deux dimensions
    # comportant len(Madj) et len(Madj[0]) colonnes (en verité len(Madj) = len(Madj[0]) car la matrice est carrée)
    for i in range(len(Madj)):
        for j in range(len(Madj[0])):
            if Madj[i][j] == 1: ## s'il existe un arc du sommet numéroté i vers le sommet numéroté j
                u, v = DNS[i], DNS[j] ## on détermine le noms des sommets correspondants
                DLadj[u].append(v)    ## on ajoute l'arc au dictionnaire des listes d'adjacence
    return DLadj
```

**Q17.a.**

*Par facilité, on désigne dans cette question les sommets par leur numéro.*

* initialisation

Pour $k=1$, $M^1_{i, j}$ vaut zéro si l'arc $i\rightarrow j$ existe dans le graphe et un sinon. $M^1_{i, j}$ donne donc bien le nombre d'arcs depuis le sommet $i$ vers le sommet numéroté $j$, soit le nombre de chemins de longueur $1$, d'origine $i$ et d'extrémité $j$.

* hérédité

Supposons que pour un $k$ donné, $k\geqslant 1$, $M^k_{i, j}$ est égal au nombre de chemins de longueur $k$, d'origine  $i$ et d'extrémité $j$.  
   
Un chemin de longueur $k+1$ entre un sommet $i$ et un sommet $j$ peut se décomposer de façon unique en un chemin de longueur $k$ joignant le sommet $i$ à un sommet $q$ (les $k+1$ arcs composant le chemin) et un chemin de longueur $1$ entre ce sommet $k$ et le sommet $j$ (le dernier arc du chemin).  
Et, pour une valeur de $q$ donnée (*i.e.* pour un avant-dernier sommet du chemin de longueur $k+1$ donné), chaque chemin de longueur $k$ entre le sommet $i$ et le sommet $q$ peut être prolongé par n'importe quel arc (il y en a un au plus !) reliant le sommet $q$ au sommet $j$, aussi le nombre de chemins de longueur $k+1$ ayant pour avant-dernier sommet le sommet $q$ est égal au produit du nombre de chemins de longueur $k$ reliant $i$ à $q$ et au nombre d'arcs reliant $q$ à $j$ (zéro ou un).
  
Si on note ${\mathcal C}_{k, a, b}$ l'ensemble des chemins de longueur $k$ entre le sommet $a$ et le sommet $b$, et $N_{k, a, b}$ son cardinal, alors, en fonction des deux observations précédentes :
$$
{\mathcal C}_{k, i, j} = \bigcup_{q=0}^{n-1} \left({\mathcal C}_{k, i, q}\times {\mathcal C}_{1, q, j} \right)
$$
où l'union sur les différentes valeurs possibles de $q$ est disjointe. 
Et donc on a les égalités de cardinaux suivantes :
$$
\left|{\mathcal C}_{k+1, i, j}\right|
= \sum_{q=0}^{n-1} \left|{\mathcal C}_{k, i, q}\right| \times \left|{\mathcal C}_{1, i, j}\right|.
$$
Or, selon les règles du calcul matriciel :
$$
M^{k+1}_{i, j} = \sum_{q=0}^{n-1} M^k_{i, q} \times M^1_{q, j}, 
$$
et, en vertu de l'hypothèse de récurrence et du résultat établi au rang $1$, 
$$
\left|{\mathcal C}_{k, i, q}\right| = M^k_{i, q} \quad {\rm et} \quad \left|{\mathcal C}_{1, q, j}\right| = M^1_{q, j}
$$
donc, on peut conclure que :
$$
\left|{\mathcal C}_{k+1, i, j}\right| = \sum_{q=0}^{n-1} M^k_{i, q} \times M^1_{q, j} = M^{k+1}_{i, j},
$$
ce qui établit l'hérédité et permet de conclure que la propriété à démontrer est vraie pour tout $k\geqslant 1$.

**Q17.b.**

Implémenter la fonction puissance consiste à implémenter la multiplication matricielle de deux matrices, à l'aide de la formule de calcul du coefficient $(i,j)$ du produit de deux matrices $A$ et $B$ de tailles compatibles, *i.e.* telles que $A$ soit de format $(n, p)$ et $B$ de format $(q, m)$ avec $p=q$ :
$$\forall (i, j) \in [\![1, n]\!] \times [\![1, m]\!], 
\quad
(AB)_{i, j} = \sum_{r=1}^{p} a_{i, r}.b_{r, j},
$$
puis à calculer le produit $M^k$.  

On peut définir une fonction auxiliaire calculant le produit de deux matrices $A$ et $B$ représentées par des listes de listes `A` et `B` :

```python
def produit(A, B):
    n, p = len(A), len(A[0]) # nombre de lignes, nombre de colonnes de A
    q, m = len(B), len(B[0]) # nombre de lignes, nombre de colonnes de B
    assert len(A[0]) == len(B) ## <-- on s'assure que le produit AB est défini
    ## on créé une matrice de zéros de même taille que le produit AB
    C = [[0] * m for i in range(n)]
    ## on calcule les coefficients de C = AB avec la formule rappelée
    ## on utilise la numérotation des lignes et des colonnes induite par l'implémentation en Python (de 0 à n - 1 ou m - 1)
    for i in range(n):
        for j in range(m):
            s = 0
            for r in range(p):
                s += A[i][r] * B[r][j]
            C[i][j] = s
    ## on renvoie la matrice créée
    return C
```

La complexité de l'algorithme est en $O(n\times m\times p)$ (on répète $nm$ fois le calcul d'une somme de $p$ termes), pour des matrices carrées d'ordre $n$, il est en $O(n^3)$).

On peut alors implémenter le calcul de la puissance $k^{\rm ième}$ d'une matrice carrée, pour $k\geqslant 1$ :

```python
def puissance(M, k):
    assert len(M) == len(M[0]) and k >= 1
    C = M
    for _ in range(k - 1): ## on a utilisé ici un indice de boucle anonyme, _, car l'indice de boucle n'est pas utilisé
        C = produit(C, M)
    return C
```

Pour une matrice carré d'ordre $n$, la complexité de l'algorithme est en $O(k\times n^3)$ (on répète $k-1$ fois un appel à la fonction `produit` de complexité $O(n^3)$).

On peut améliorer la complexité en utilisant un algorithme d'***exponentiation rapide***, vu en première année : la complexité sera alors en $O\left(\log_2(k)\times n^3\right)=O\left(n^3\ln k\right)$.

On peut, pour chaque sommmet $s$, tester pour chacun de ses voisins, $v$, si le sommet $s$ est dans la liste d'adjacence de $v$ (et interrompre le parcours si c'est le cas).

```python
def est_non_orienteLA(DLadj):
    for s, Ladj in DLadj.items():
        for v in Ladj:
            if s not in DLadj[v]: # test de complexité dans le pire cas proportionnel aux nombres de voisins de v
                return False
    return True
```

La complexité dans le pire cas (lorsque le graphe est non orienté) est en
$$\sum_{i=0}^{n-1} \left(\sum_{j=1}^{m_i} O\left(m_{i, j}  \right)\right)
= O\left(\sum_{i=0}^{n-1} \left(\sum_{j=1}^{m_i} m_{i, j}  \right)\right),$$
où $m_{i_j}$ est le nombre de voisins du $j^{\rm ème}$ voisin du sommet $i$ dont on teste si l'un des voisins est le sommet $i$.

Une valeur $m_{i, j}$ est égal au degré sortant d'un sommet voisin de $v$ et peut être majoré par le maximum des degrés sortant pour les sommets du graphe, lui-même majoré par le nombre de sommets du graphe.  
On peut donc majorer la complexité dans la pire cas de la façon suivante :
$$
T(n, m)
= O\left(\sum_{i=0}^{n-1} \left(\sum_{j=1}^{m_i} m_{i, j}  \right)\right)
= O\left(\sum_{i=0}^{n-1} \left(\sum_{j=1}^{m_i} n  \right)\right)
=O\left(m \times n\right).$$

Cette complexité est à comparer à celle de la fonction du **18.a.**, en $O\left n^2 \right)$.   
Comme pour un graphe d'ordre $n$, le nombre d'arcs est majoré par $n^2$, la complexité de la fonction du **18.b.** est potentiellement en $O\left(n^3\right)$, *a priori* moins bonne que celle du **18.a**,mais ceci est à tempérer par estimation, dans les graphes considérés du rapport entre la taille et l'ordre.

**Q20.a**

En utilisant la matrice d'adjacence, on peut déterminer la liste des sommets à la distance $1$ d'un sommet $i$, en parcourant la ligne d'indice $i$ de la matrice (on détermine ainsi la liste d'adjacence de $i$).  
Puis, pour déteminer les sommets à la distance $2$ de $i$, on peut explorer le voisinage des sommets situés à distance $1$ de $i$, en excluant les sommets déjà trouvés comme étant à distance $1$ de $i$.  
De façon générale, on trouvera les sommets à distance $d+1$ de $i$ en explorant le voisinages des sommets à distance $d$ de $i$ et en excluant de la recherche les sommets déjà trouvés situés à une distance inférieure ou égale à $d$.

On ne le fera pas ici, mais il y a aurait un gain de complexité temporelle à déterminer les listes d'adjacence de tous les sommets (pour éviter de parcourir à chaque fois une ligne de longueur $n$ pour rechercher les voisins des sommets à distance $d$ de $i$), ce gain de complexité temporelle se ferait au prix d'un espace en mémoire permettant de stocker le dictionnaire des listes d'adjacence du graphe, en $O(m)$, majoré par $O(n^2)$).  
Ceci nous rapprocherait de l'algorithme de parcours en largeur.

Afin d'exclure de la recherche les sommets déjà trouvés à une distance inférieure strictement inférieure à $d$, on maintiendra un tableau $T$ des distances déjà trouvées.  
$T[j]$ pour tout $j$ sera initialisé à `None`, et prendra la valeur $d$ pour le plus petit $d$ tel que `Madj[i][j] > 0`. 

**On notera bien qu'il ne peut exister de sommets à distance $k > n$ de $i$, car un plus court chemin de $i$ à un autre sommet ne peut comporter plusieurs fois le même sommet, donc au pire un tel chemin comporte $n+1$ sommets (s'il part d'un sommet et y revient, en passant par chacun de tous les autres sommets, c'est alors un  circuit) et sa longueur est inférieure au égale à $n$.**

```python
def voisins_a_distance_donnee(Madj, i, k, affichage = False):
    assert 1 <= k <= len(Madj)
    n = len(Madj)
    T = [None for k in range(n)]
    ## on initialise avec la liste des voisins à distance 1
    voisins = []
    for j in range(n):
        if Madj[i][j] == 1:
            voisins.append(j)
            T[j] = 1 # on note dans T les voisins à distance 1
    if affichage: print("\td =", 1, "voisins :", voisins, "; T =", T)
    ## recherche des voisins à distance d, 2 <= d <= k
    for d in range(2, k + 1):
        voisins_suivants = []
        for v in voisins:
            for j in range(n):
                if Madj[v][j] == 1 and T[j] == None:
                    voisins_suivants.append(j)
                    T[j] = d # on note dans T les voisins à distance d
        voisins = voisins_suivants
        if affichage: print("\td =", 1, "voisins :", voisins, "; T =", T)
    return voisins
```

On parcourt initialement la ligne $i$ de la matrice, puis, pour chaque distance examinée, $d$ comprise entre $2$ et $k$, on parcourt $n_{d-1}$ lignes de la matrice d'adjacence, où $n_{d-1}$ est égal au nombre de sommets à distance $d-1$ de $i$, on peut donc majorer la complexité par :
        $$T(n, m, k) \leqslant O(n) + \sum_{d=1}^{k -1} O(n_d \times n) 
        \leqslant O(n) + n \times \sum_{d=1}^{k -1} O(n_d)
        \leqslant O\left(n \times \sum_{d=1}^{k -1}n_d\right).$$
Or $\sum_{d=1}^{k -1} n_d \geqslant n$ (le nombre de sommets à distance inférieure strictement à $k$ ne peut excéder l'ordre du graphe), donc on peut majorer la complexité par $O\left(n^2\right)$. 

Puisque $m\leqslant n^2$, on trouve une complexité comparable, à celle d'un parcours en largeur, qui est en $O\left(n + m\right)$, moins bonne si le nombre d'arêtes, $m$, est *très inférieure à $n^2$*.

On adopte une ici une démarche élémentaire et brutale, consistant à calculer les puissance de la matrice d'adjacence $M$, et à stopper leur calcul dès qu'une valeur non nulle est trouvée sur la diagonale : si un coefficient diagonal de $M^k$ est non nul, c'est qu'il existe un chemin de longueur $k$ d'un sommet vers lui-même.  
Le calcul des puissances peut être arrêté dans tous les cas lorsque $k=n$, car on a vu à la question **20.** qu'un circuit (et aucun autre chemin) ne peut être de longueur strictement plus grande que $n$.

La complexité sera médiocre, car chaque itération coûte une multiplication en $O(n^3)$, tandis que chaque examen coûte $O(n)$.  
La complexité dans le pire cas sera donc en $O(n^4)$ et donc **extrèmement mauvaise**.

On s'autorise de tester que la diagonale comporte un coefficient non nul en calculant la somme des termes diagonaux,
ce qui n'augmente pas la complexité du test qui reste, cette fois dans tous les cas, en $O(n)$.

On réutilise ou on s'inspire des fonctions de la question **17.**

La fonction ci-dessous renvoie la longueur du circuit trouvé, 0 s'il n'y en a pas

```python
def existe_cricuit(Madj):
    n = len(Madj)
    C = Madj
    if sum([C[i][i] for i in range(n)]) > 0:
        return True, 1 ## il existe une boucle (circuit de longueur 1)
    #
    for k in range(2, n + 1):
        C = produit(C, Madj) ## une nouvelle matrice est construite par la fonction produit, ce qui assure que Madj
                             ## n'est pas modifiée au fil des itérations, en dépit de l'instruction initiale C = Madj
        if sum([C[i][i] for i in range(n)]) > 0:
            return True, k ## il existe une boucle (circuit de longueur p)
    #
    return False, 0
```
