* deuxième version

Le deuxième version demandée ne doit pas faire appel à la fonction `posmin`. L'implémenter consiste à intégrer le code de la fonction `posmin` dans le corps de la fonction `tri_selection1`.  
Il est possible de copier/coller le code de la fonction `posmin` en lieu et place de l'instruction `posmini = posmin(L, i)` dans le corps de la fonction `tri_selection1`, mais il faut alors prêter attention à ce qu'il n'y ait pas de conflit de noms entre les variables locales du corps de la fonction `posmin` et celles du corps de la fonction `tri_selection1` : en particulier, veiller à ce que les indices des deux boucles `for` n'aient pas le même nom.

```python
def tri_selection2(L):
    for i in range(len(L) - 1):
        # recherche d'une position du minimum des valeurs en position i et au-delà
        jmin = i
        for j in range(i + 1, len(L)):
            if L[j] < L[jmin]:
                jmin = j
        # échange de la valeur en position jmin (minimum des valeurs de L[i:]) et de la valeur en position i
        tmp = L[i]
        L[i] = L[jmin]
        L[jmin] = tmp
```

<i><u>Note</u> :</i> l'échange de valeurs entre les positions `jmin` et `i` réalisé ci-dessus avec la méthode classique utilisant une variable auxiliaire, `tmp`, et trois instructions d'affectation, peut se faire aussi grâce à une unique instruction d'affectation multiple `L[i], L[jmin] = L[jmin], L[i]`.

```python
def insere(L, i):
    tmp = L[i] 
    j = i
    while 0 < j and L[j - 1] > tmp:
        L[j] = L[j - 1]
        j = j - 1
    L[j] = tmp
```

## Exercice 3 &emsp; Tri rapide

L'algorithme de tri rapide est un algorithme de type "diviser pour régner" (de même que le tri-fusion déjà traité).

On en examine ici une implémentation récursive.

Le principe de cet algorithme consiste à :  
1. choisir, arbitrairement, une position ***pivot*** dans la liste à trier (ici systématiquement la position zéro). La valeur située à la position *pivot* sera nommée ***valeur-pivot*** ;  
2. construire la liste des valeurs, ***situées à une autre position que le pivot*** (ici, situées en position 1 et au-delà), qui sont inférieures ou égales à la valeur-pivot - on appelera &laquo;$\,$*listes de plus petits*$\,$&raquo; les listes ainsi créées,  
et la  la liste des valeurs, ***situées à une autre position que le pivot*** (ici, situées en position 1 et au-delà), qui sont strictement supérieures à la valeur-pivot - on appelera &laquo;$\,$*listes de plus grands*$\,$&raquo; les listes ainsi créées ;  
3. itérer le procédé sur toutes les nouvelles listes ainsi construites, tant que ces listes sont de longueur supérieures ou égales à 2 ;  
4. en partant des listes de longueur inférieure ou égale à 1 obtenues, recombiner les listes de plus petits et de plus grands en des listes triées en réintégrant à chaque fois la valeur du pivot utilisé lors de la scission.

```python
def tri_rapide(L):
    ## traitement des cas de base
    if len(L) <= 1:
        return L[:] ## on renvoie une copie de la liste en entrée, c'est une version triée de L si len(L) <= 1
    ## traitement du cas général (cas où len(L) >= 2)
    # on recopie toutes les valeurs de la liste L, en les répartissant en une valeur-pivot,
    # une liste des valeurs autre que la valeur-pivot et inférieures ou égales à la valeur-pivot, et
    # une liste des valeurs autre que la valeur-pivot et supérieures strictement à la valeur-pivot
    pivot, pluspetits, plusgrands = partage(L)
    # tri des listes de "plus petits" et de "plus grands" grâce à des appels récursifs
    petits_tries, grands_tries = tri_rapide(pluspetits), tri_rapide(plusgrands)
    # combinaison des listes triées obtenues et de la valeur utilisée comme valeur-pivot
    # en une liste triée (version triée de la liste L en entrée)
    return petits_tries + [pivot] + grands_tries
```

```python
def tri_denombrement2(L):
    ## premier parcours de L :
    ##         détermination du maximum, M, des valeurs dans L, et du dictionnaire des nombres d'occurrences
    M, d = L[0], {} # <--- initialisation du maximum M, et du dictionnaire d
    for val in L:
        d[val] = 0
        if val > M:
            M = val
    ## deuxième parcours de L :
    ##         comptage des occurrences des différentes valeurs dans L
    for val in L:
        d[val] += 1
    ## construction d'une version triée de L à partir des nombre d'occurrences stockés dans d
    Lt = []
    for val in range(M + 1):
        if val in d.keys():
            Lt.extend([val] * d[val])
    return Lt
```

**Q6** Discuter la complexité en temps et en espace de ces deux implémentations du tri par dénombrement.

Ces questions seront abordées plus précisément, avec les outils adaptés, au moment du TP12.

* complexité en espace (complexité spatiale) :

***Il s'agit ici d'estimer la quantité de mémoire nécessaire pour exécuter un algorithme donnant la solution à un problème (ici produire une version triée d'une liste d'entiers naturels), en fonction de la taille de l'instance (ici le nombre d'éléments de la liste à trier).***

La mise en œuvre du <u>premier algorithme</u> de tri par dénombrement d'une liste `L` (implémenté par la fonction `tri_denombrement`), hormis les variables locales servant à stocker des valeurs entières (indices de boucle, variable `M`, ...), nécessite la création d'un tableau des occurrences, de taille $M+1$ et d'une liste de même taille que la liste en entrée (version triée de L), *i.e* de taille $n={\rm len}L$.  

La **complexité en espace du premier algorithme** (quantité d'espace nécessaire en mémoire pour exécuter l'algorithme) est donc (environ) proportionnelle à $M+n$, où $n={\rm len}L$ et $M=\max(L)$.

La mise en œuvre du <u>second algorithme</u> de tri par dénombrement d'une liste `L` (implémenté par la fonction `tri_denombrement2`), hormis les variables locales servant à stocker des valeurs entières (indices de boucle, variable `M`, ...), nécessite la création d'un dictionnaire des occurrences, de taille $n'$ où $n'$ est le nombre de valeurs distinctes dans `L`, et d'une liste de même taille que la liste en entrée (version triée de L), *i.e* de taille $n={\rm len}L$.  

La **complexité en espace du second algorithme** (quantité d'espace nécessaire en mémoire pour exécuter l'algorithme) est donc (environ) proportionnelle à $n'+n$, où $n={\rm len}L$ et $n'$ est le nombre de valeurs distinctes dans $L$ (on notera donc que $n'\leqslant n$).

*Dans le cas où le maximum des valeurs entières positives dans $L$ peut être considéré comme très supérieur au nombre de valeurs dans $L$, la complexité en espace du second algorithme est donc (nettement) meilleure.*

* complexité en temps (complexité temporelle) :

***Il s'agit ici d'estimer le temps nécessaire pour exécuter un algorithme donnant la solution à un problème (ici produire une version triée d'une liste d'entiers naturels), en fonction de la taille de l'instance (ici le nombre d'éléments de la liste à trier).***

<b><i><u>Note</u> :</i></b> $\;$Il est d'usage de parler de la complexité (sans plus préciser) d'un algorithme lorsque l'on souhaite parler de la complexité temporelle de cet algorithme.

*La méthode usuelle retenue pour estimer la complexité temporelle est de dénombrer le nombre d'opérations élémentaires à effectuer (affectations, comparaisons de valeurs, opérations arithmétiques sur des entiers de taille &laquo;$\,$raisonnable$\,$&raquo;, sur des flottants ou des booléens, ajout d'une valeur à une liste à l'aide de la méthode `.append`, ajout d'un couple (clé, valeur) dans un dictionnaire, accès à une valeur dans une liste à l'aide de son indice ou à une valeur dans un dictionnaire à l'aide de la clé associée, ...).*

La mise en œuvre du <u>premier algorithme</u> de tri par dénombrement d'une liste `L` (implémenté par la fonction `tri_denombrement`), peut être décomposé en :<br/>
$\quad \star\;$ la recherche du maximum, $M$, de complexité (temporelle) proportionnelle au nombre d'éléments dans la liste, $n={\rm len}(L)$ ;  
$\quad \star\;$ la construction d'un tableau des occurrences de taille $M+1$, de complexité proportionnelle à $M$ ;  
$\quad \star\;$ un parcours de `L` pour mettre à jour le tableau des nombres d'occurrences, de complexité proportionnelle à $n$ ;  
$\quad \star\;$ la construction d'une version triée de `L` de complexité proportionnelle au nombre de valeurs dans `L`, *i.e.* proportionnelle à $n$.  
Ainsi, l'algorithme, résumé à ces quatre phases, comporte un ensemble d'opérations de complexité proportionnelle à $M$ et trois ensembles d'opérations de complexité proportionnelle à $n={\rm len}(L)$.

La **complexité (temporelle) du premier algorithme** est donc (environ) proportionnelle à $M+3n$, où $n={\rm len}L$ et $M=\max(L)$.

La mise en œuvre du <u>second algorithme</u> de tri par dénombrement d'une liste `L` (implémenté par la fonction `tri_denombrement2`), peut être décomposé en :<br/>
$\quad \star\;$ la recherche du maximum, $M$ et l'initialisation du dictionnaire `d`, de complexité (temporelle) proportionnelle au nombre d'éléments dans la liste, $n={\rm len}(L)$ ;   
$\quad \star\;$ un parcours de `L` pour mettre à jour le dictionnaire des nombres d'occurrences, de complexité proportionnelle à $n$ ;  
$\quad \star\;$ une boucle dont l'indice varie de $0$ à $M$ pour tester si chaque valeur comprise entre $0$ et $M$ est ou non une clé du dictionnaire `d`, et au cours de laquelle sont effectués $M+1$ tests `val in d.keys()` nécessitant chacun au pire une comparaison de la valeur `val` avec les $n'$ clés du dictionnaire `d`, soit au total un nombre d'opérations élémentaires (au pire) proportionnel au produit $M\times n'$ où $n'$ est égal au nombre de vlaeurs distinctes dans `L` (nombre de clés du dictionnaire `d`) ;  
$\quad \star\;$ la construction d'une version triée de `L` de complexité proportionnelle au nombre de valeurs dans `L`, *i.e.* proportionnelle à $n$.  
Ainsi, l'algorithme, résumé à ces quatre phases, comporte un ensemble d'opération de complexité proportionnelle à $Mn'$, un ensemble d'opération de complexité proportionnelle à $M$ (dont la complexité pourrait donc être agrégée à la complexité de l'ensemble d'opérations précédent), et deux  ensembles d'opérations de complexité proportionnelle à $n={\rm len}(L)$.

La **complexité (temporelle) du second algorithme** est donc (environ) proportionnelle à $M(1+n')+2n$, où $n={\rm len}L$, $n'$ est le nombre de valeurs distinctes dans `L`,  et $M=\max(L)$.

*La complexité temporelle du second algorithme est donc potentiellement plutôt moins bonne que celle du premier algorithme si $Mn'$ peut être considérée comme très supérieure à $M$.*

* comparaison empirique des temps d'exécution des deux algorithmes de tri par dénombrement pour des listes aléatoires d'entiers de longueur $n$, pour $n$ compris entre $10000$ et $100000$, par pas de $10000$.

$\quad\quad\star$ pour des listes dans lesquelles environ $63$ % des valeurs de l'intervalle $[\![0, \max(L)]\!]$ sont représentées :

```python
from time import perf_counter

pmax = 10
N, F, T1, T2 = [], [], [], []
#for p in range(pmax + 1):
#    n = 2 ** p
for n in range(10000, 100000 + 1, 10000):
    N.append(n)
    L = [randint(1, n) for i in range(n)]
    F.append(len(set(L)) / (max(L) + 1))
    tmin = float('inf')
    for _ in range(10):
        t0 = perf_counter()
        tri_denombrement(L)
        t1 = perf_counter() - t0
        if t1 < tmin:
            tmin = t1
    T1.append(tmin)
    tmin = float('inf')
    for _ in range(10):
        t0 = perf_counter()
        tri_denombrement2(L)
        t1 = perf_counter() - t0
        if t1 < tmin:
            tmin = t1
    T2.append(tmin)
print('Proportion de valeurs de l\'intervalle [0, max(L)] représentées :', sum(F) / len(F))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(N, T1,
            label = 'denombrement1')
ax.plot(N, T2,
            label = 'denombrement2')
ax.set_xlabel('temps d\'exécution')
ax.set_ylabel('longueur de la liste')
ax.legend()
plt.show()
```

**On constate empiriquement que les complexités sont comparables et semble être proportionnelles à la taille de la liste.**
