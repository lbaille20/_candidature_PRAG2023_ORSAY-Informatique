**Q2 $\quad$ Degrés, entrant et sortant, d'un sommet**

Pour un graphe orienté, représenté comme dans l'exercice 1, écrire quatre fonctions `degre_sortantM(M, i)`, `degre_entrantM(M, i)`, `degre_sortantLA(LA, i)`, `degre_entrantLA(LA, i)`, renvoyant le degré entrant ou sortant du sommet $i$, pour les deux premières en supposant que le graphe est représenté par une matrice d'adjacence, `M`, et pour les deux dernières en supposant que le graphe est représenté par la liste des listes d'adjacences de ses sommets, `LA`.  
On donnera pour chacune de ces quatre fonctions sa complexité dans le pire cas, en fonction de l'ordre et/ou de la taille du graphe.

```python
def degre_sortantM(M, i):
    s = 0
    for j in range(len(M[j])):
        s += M[i][j]
    return s
```

Complexité : $O(|G|)$.

```python
def degre_entrantM(M, i):
    s = 0
    for k in range(len(M)):
        s += M[k][i]
    return s
```

Complexité : $O(|G|)$.

```python
def degre_sortantLA(LA, i):
    return len(LA[i])
```

Complexité : $O(1)$.

```python
def degre_entrantLA_v0(LA, i):
    s = 0
    for k in range(len(LA)):
        if i in LA[k]:
            s += 1
    return s
```

Il n'est *a priori* pas autorisé de recourir au test `in` pour la recherche d'une valeur dans une liste (*cf.* annexe A du programme d'informatique de tronc commun), aussi on implémentera explicitement cette recherche :

```python
def degre_entrantLA(LA, i):
    s = 0
    for k1 in range(len(LA)):
        for k2 in range(len(LA[k1])):
                if LA[k1][k2] == i:
                    s += 1
    return s
```

Une première analyse rapide dans laquelle on majore les longueurs `len(LA[k1])` par $|A|$, donne une complexité en $O(|G|\times |A|)$, par le calcul 
$$
T(G)= \sum_{k_1=0}^{|G|-1} \left(c_1 +\sum_{k_2=0}^{d_+(k_1)-1} c_2\right)
\leqslant
\sum_{k_1=0}^{|G|-1} \left(c_1+\sum_{k_2=0}^{|A|-1} c_2\right)
= 
|G|\times \left(c_1 + |A| \times c_2\right),
$$
où $c_1$ et $c_2$ sont des constantes.  
Mais, considérant que $\sum_{s \in {\mathcal S}} d_+(s) = |A|$ (la somme des degrés sortants dans un graphe orienté est égale au nombre d'arcs du graphe - tout comme la somme des degrés entrants), on a en fait :
$$
T(G)= \sum_{k_1=0}^{|G|-1} \left(c_1+ \sum_{k_2=0}^{d_+(k_1)-1} c_2\right)
= \sum_{k_1=0}^{|G|-1}\left(c_1+ c_2.d_+(k_1)\right)
= \sum_{k_1=0}^{|G|-1}c_1 + c_2\sum_{k_1=0}^{|G|-1}d_+(k_1)
= c_1.|G| + c_2. |A|.$$

On peut donc conclure sur une complexité en $O(|G| + |A|)$.

S'il existe un circuit dans un graphe orienté $G$, on peut réduire ce circuit à un circuit *élémentaire*, *i.e.* ne comportant pas plus d'une fois le même sommet (hormis le sommet situé aux deux extrémités du circuit), et cond ne chercher que l'existence de circuits de longueur inférieure ou égale à $n+1$ où $n$ est l'ordre du graphe.

En conséquence, le calcul de la puissance $(n+1)^{\rm ème}$ de $M$ ($M$ matrice d'adjacence du graphe, $n$ ordre du graphe) permet de déterminer si un circuit existe dans le graphe $G$, en examinant s'il y a un coefficient non nul sur sa diagonale.

Afin de minimiser la complexité de l'algorithme, on pourra tester la diagonale des matrices $M^k$ pour chaque $k$ entre $0$ et $n+1$ afin d'interrompre la recherche à la première détection d'un circuit

On peut alors proposer la fonction suivante (où l'on s'autorise à recourir à la fonction prédéfinie `sum`) :

```python
def existe_circuit(M):
    import numpy as np
    n = len(M)
    MT1 = np.array(M)
    MTk = np.array(M)
    if sum([MTk[i, i] for i in range(n)]) > 0:
        return True, 1
    for p in range(2, n + 2):
        MTk = np.dot(MTk, MT1)
        if sum([MTk[i, i] for i in range(n)]) > 0:
            return True, p
    return False
```

* analyse de complexité

```python
def existe_circuit(M):
    import numpy as np
    n = len(M)
    MT1 = np.array(M) ## O(n^2)
    MTk = np.array(M) ## O(n^2)
    if sum([MTk[i, i] for i in range(n)]) > 0: ## O(n)
        return True, 1
    for p in range(2, n + 2):
        MTk = np.dot(MTk, MT1) ## O(n^2)
        if sum([MTk[i, i] for i in range(n)]) > 0: ## O(n)
            return True, p
    return False
```

S'il existe une boucle dans le graphe (meilleur cas), la complexité de l'algorithme est donnée par :
    $$T(n) = O(n^2) +   O(n^2) +   O(n) =O(n^2).$$  
S'il n 'existe aucun circuit dans le graphe (pire cas), la complexité de l'algorithme est donnée par :
    $$T(n) = O(n^2) +   O(n^2) +   O(n) + \sum_{p=2}^{n+1} O(n^2) +   O(n)
    = O(n^2) + n \times O(n^2) =O(n^3).$$
