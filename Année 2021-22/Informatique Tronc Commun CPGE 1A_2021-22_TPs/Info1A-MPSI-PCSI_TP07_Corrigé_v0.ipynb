## Introduction

Une fonction récursive est une *fonction qui s'appelle elle-même* : cette possibilité, pour surprenante qu'elle puisse être au premier abord, est offerte à toute fonction Python, cependant, mal utisée, elle peut ne pas produire le résultat attendu et surtout produire des fonctions qui ne termineront pas (à l'instar d'une boucle `while` qui ne termine pas.  

Ce TP n'aborde pas les questions théoriques de la bonne construction d'une fonction récursive ou de la terminaison de tout appel à une fonction récursive donnée, mais essaie de faire percevoir le fonctionnement de la récursivité et de donner des éléments pour construire une fonction récursive correcte dans des situations bien balisées.  

Ce TP termine sur l'implémentation d'algorithmes de tri, dont on propose une implémentation impérative (utilisant des boucles) et une implémentation récursive.

## Exercice 1 fonction récursive produisant des affichages

On considère la fonction `f` suivante :

```python
def f(message, n):
    for i in range(n):
        print(message)
```

Sans surprise, cette fonction aura pour effet d'afficher $n$ fois un message passé en argument sous la forme d'une chaîne de caractères :

```python
f('bonjour', 3)
```

Le rôle de la boucle `for` dans le corps de la fonction `f` est de répéter l'exécution de l'instruction `print(message)`.  
On conçoit qu'un appel répété $n$ à une fonction `g` ne contenant qu'une instruction `print(message)` et aucune boucle, produirait le même effet... C'est ce que l'on peut réaliser avec la fonction récursive `g` suivante :

```python
def g(message, n):
    if n > 0:
        print(message)
        g(message, n - 1)
```

* test

```python
g('bonjour', 3)
```

**Q1** Quel(s) effet(s) (d'affichage) produisent les appels `g('bonjour', 0)` et `g('bonjour', 1)` ?

***Q1 corrigé***

L'appel `g('bonjour', 0)` ne produit aucun effet d'affichage, car la valeur de $n$ ($n=0$) n'est pas strictement positive.  L'appel termine donc sans qu'aucune instruction, `print(message)` ou `g(message, n - 1)`, soit exécutée. L'appel termine et la fonction renvoie la valeur `None`.<br/>
L'appel `g('bonjour', 1)` produit un affichage du mot 'bonjour', car $n$ vaut 1 dans cet appel, et donc le test `n > 0` est évalué à `True` et l'instruction `print(message)` est exécutée, de même que l'appel `g(message, n - 1)` dans lequel le second argument vaut $1-1=0$, cet appel, *cf.* ci-avant, termine sans rien afficher et en renvoyant la valeur `None`. L'appel `g('bonjour', 1)` termine alors, en renvoyant lui-aussi la valeur `None`. 

```python
res0 = g('bonjour', 0)
```

```python
print(res0)
```

```python
res1 = g('bonjour',1)
```

```python
print(res1)
```

**Q2** Compter les appels à la fonction `g` qui découlent de l'appel `g('bonjour', 3)`. Quelle(s) est(sont) la(les) valeur(s) du parmètre $n$ pour cet(ces) appel(s) ?  
Qu'est-ce qui, dans le code de la fonction `g`, assure que, suite à l'exécution d'un appel à `g`, les appels récursifs soient en nombre fini ? Quelle est la condition à imposer sur le paramètre $n$ pour qu'il en soit ainsi ?

***Q2 corrigé***

`g('bonjour', 3)` déclenche trois appels récursifs, pour lesquels le paramètre $n$ a pour valeurs, successivement : $2, 1$ et $0$.<br/>
C'est la décroissance de $n$ au fil des appels récursifs qui assure que les appels récursifs sont en nombre fini : un appel `g('bonjour', n)` engendre un unique appel `g('bonjour', n - 1)`, qui engendre un unique appel `g('bonjour', n - 2)`, *etc.* jusqu'à un dernier appel `g('bonjour', 0)`.<br/> 
**Pour qu'il en soit ainsi, il est nécessaire que la valeur initiale de $n$ soit positive !** (il y a alors $n$ appels récursifs).<br/>
Pour une valeur initiale de $n$ strictement négative, les appels récursifs s'enchaînent à l'infini et un appel `g('bonjour', n)` ne termine pas.

## Exercice 2 fonction récursive implémentant l'application d'une formule de récurrence

On s'intéresse maintenant à une fonction récursive ne produisant cette fois aucun effet, mais renvoyant une valeur (autre que `None`), ici une valeur entière.  
On considère la fonction suivante, traditionnellement utilisée comme premier exemple de fonction récursive :

```python
def fact(n):
    if n == 0:
        return 1
    return n * fact(n - 1)
```

Sans trop de surprise, on peut constater que cette fonction implémente, pour une entrée dans le domaine des entiers naturels, la fonction mathématique **factorielle** :

```python
fact(0)
```

```python
fact(5)
```

*Note :* la fonction factorielle est implémentée dans le module `math` sous le nom `factorial`

```python
from math import factorial
factorial(5)
```

On constate dans l'implémentation récursive présentée ci-avant

```python
def fact(n):
    if n == 0:
        return 1
    return n * fact(n - 1)
```

qu'il est implementé, *via* le branchement conditionné par le test `n == 0` et l'instruction `return 1`, que la valeur de $0!$ est 1. L'exécution de l'instruction `return 1` termine dans ce cas la fonction.  
Et pour tout $n\geqslant 1$, on peut observer que le résultat, entier, de l'appel récursif `fact(n - 1)` est utilisé dans le calcul de $n!$.

En résumé, on peut constater que le fonctionnement récursif permet d'implémenter le calcul de $n!$ de façon *quasi* littérale à partir de la définition par récurrence de la factorielle d'un entier $n\geqslant 0$ :

$$\left\{\begin{array}{l}
0! = 1\\
\forall n\geqslant 1, \quad n! = n \times (n - 1)!
\end{array}\right.$$

**Q1.a** Sur le même modèle, implémenter sous la forme d'une fonction récursive `u(n)`, le calcul du terme de rang $n$ de la suite définie par récurrence par 
$$\left\{\begin{array}{l}
u_0 = 1\\
\forall n\geqslant 1, \quad u_n = 3 u_{n-1} + 2.
\end{array}\right.$$

***Q1.a corrigé***

```python
def u(n):
    if n == 0 :
        return 1
    return 3 * u(n - 1) + 2
```

**Q1.c** Donner une implémentation itérative (utilisant une boucle) du calcul du terme de rang $n\geqslant 0$ de la suite $u$ sous la forme d'une fonction `u_it`, et vérifier les résultats de la question **Q1.b**.

***Q1.c corrigé***

On utilise une variable locale, `u`, donc la valeur est actualisée à chaque itération.  
Avant toute itération, la valeur de `u` est $u_0$.  
À la fin de l'itération d'indice $k$, la valeur de `u` est $u_{k+1}$.  
Lorsque la boucle `for` termine, la valeur de `u` est $u_n$.

**Q2** Définir une fonction récursive, `p2`, de paramètre $n$ et renvoyant la valeur de $2^n$.  
***On s'appuyera sur la définition par récurrence d'une suite géométrique de premier terme $u_0$ et de raison $q$.***

***Q2 corrigé***

La définition par récurrence d'une suite, $(u_n)_{n\geqslant 0}$, géométrique de premier terme $u_0$ et de raison $q$ est :
$$\left\{\begin{array}{l}
u_0 \in \mathbb R\\
\forall n\geqslant 0, \quad u_{n + 1} = q.u_{n}.
\end{array}\right.$$

La suite $(u_n)_{n\geqslant 0}=(2^n)_{n\geqslant 0}$ est un cas particulier, dont la définition par récurrence est :
$$\left\{\begin{array}{l}
u_0 = 1\\
\forall n\geqslant 0, \quad u_{n + 1} = 2.u_{n}.
\end{array}\right.$$

L'écriture d'une fonction récursive, `p2`, implémentant le calcul de $2^n$ est immédiate si l'on modifie cette définition en :
$$\left\{\begin{array}{l}
u_0 = 1\\
\forall n\geqslant 1, \quad u_{n} = 2.u_{n - 1}.
\end{array}\right.$$
de sorte que la formule de récurrence donne soit la formule de calcul de $u_n$ et non de $u_{n+1}$.

```python
def p2(n):
    if n == 0 :
        return 1
    return 2 * p2(n - 1)
```

* tests

```python
for k in range(11):
    print(p2(k), end = ", ")
```

## Exercice 3 fonction récursive implémentant un parcours de liste

On rappelle le parcours d'une liste `L` à l'aide d'une boucle `while`

```python
L = ['a', 'b', 'c', 'd', 'e', 'f']

i = 0
while i < len(L):
    print(L[i])
    i += 1
```

On propose la fonction recursive suivante pour implémenter l'affichage des éléments d'une liste L.  
***On suppose que le paramètre $i$ sera toujours pris entier et positif.***

```python
def affiche_liste_rec(L, i):
    if i < len(L):
        print(L[i])
        affiche_liste_rec(L, i + 1)
```

**Q1** Quels affichages obtient-on avec les appels :
* `affiche_liste_rec([], 0)` ?  
* `affiche_liste_rec(['a', 'b', 'c', 'd', 'e', 'f'], 5)` ?  
* `affiche_liste_rec(['a', 'b', 'c', 'd', 'e', 'f'], 4)` ?  
* `affiche_liste_rec(['a', 'b', 'c', 'd', 'e', 'f'], 0)` ?

***Q1 corrigé***

Un appel `affiche_liste_rec(L, i)` avec $i\geqslant 0$ déclenche l'affichage de l'élément en position $i$ dans `L`, et l'appel récursif `affiche_liste_rec(L, i + 1)`, causant l'affichage de l'élément suivant de la liste `L`, et ainsi de suite, jusqu'à ce que $i=\rm len(L)$, auquel cas l'appel récursif termine sans rien afficher et en renvoyant la valeur `None`, ce qui provoque, de proche en proche, la fin de tous les appels antérieurs.  
  
Ainsi, de façon synthétique, un appel `affiche_liste_rec(L, i)`, pour $i\geqslant 0$, provoquera l'affichage de tous les termes de la liste extraite `L[i:len(L)]`


* vérifications

```python
affiche_liste_rec([], 0)
```

```python
affiche_liste_rec(['a', 'b', 'c', 'd', 'e', 'f'], 5)
```

```python
affiche_liste_rec(['a', 'b', 'c', 'd', 'e', 'f'], 4)
```

```python
affiche_liste_rec(['a', 'b', 'c', 'd', 'e', 'f'], 0)
```

## Exercice 4 fonction récursive implémentant un calcul de somme

On sait implémenter de façon itérative le calcul d'une somme, de la forme $\sum_{k=0}^{n} v_k$, par exemple lorsque les valeurs $v_k$ sont données par une formule de calcul en fonction de $k$, ou sont les éléments d'une liste $L=[v_0, v_2, \dots, v_n]$.  

* traitement du premier cas : si il existe une fonction $f$ telle que $v_k=f(k)$

```python
def somme_cas1(f, n):
    s = 0
    for k in range(0, n + 1):
        s = s + f(k)
    return s

## test : calcul de la somme s_10 des 10 plus petites puissances de 2 d'exposant positif
def f(k):
    return 2 ** k

s_10 = somme_cas1(f, 9)
```

```python
## vérification (en comparant avec la valeur donnée par la formule du cours de mathématiques)
s_10 == 2 ** 10 - 1
```

* traitement du second cas : les valeurs $v_k$ sont les éléments d'une liste $L=[v_0, v_2, \dots, v_n]$.

```python
def somme_cas2(L):
    s = 0
    for k in range(len(L)):
        s = s + L[k]
    return s

## test : calcul de la somme des 5 premiers entiers strictement positifs
L = [1, 2, 3, 4, 5]

S_5 = somme_cas2(L)
```

```python
## vérification (en comparant avec la valeur donnée par la formule du cours de mathématiques)
S_5 == (5 * 6) //2
```

**On cherche maintenant une implémentation récursive de ces calculs de somme**.

**Q1** Si on note, pour tout $k\in[\![0;n]\!]$, $S_n=\sum_{k=0}^{n}v_k$, donner une relation de récurrence entre deux termes consécutifs de la séquence $S_0, S_1, \dots, S_n$.

***Q1 corrigé***

Pour tout $1\leqslant i\leqslant n$, $S_i$ se calcule en ajoutant la valeur $v_i$ ) à la somme $S_{i-1}$, d'où la relation de récurrence :
$$
\forall i \in [\![1, n]\!], \quad S_i = S_{i-1} + v_i
$$
En effet :
$$
\forall i \in [\![1, n]\!], \quad 
S_{i} = \sum_{k=0}^{i}v_k = \left(\sum_{k=0}^{i-1}v_k\right) + v_{i}=  S_{i-1} + v_{i} .
$$

**Q2** À l'aide de la question **Q1**, définir deux fonctions récursives, `som1(f, n)` et `som2(L)`, implémentant, respectivement, de la calcul d'une somme de la forme $\sum_{k=0}^{n}f(k)$ et le calcul de la somme des termes d'une liste $L$.

* Tester ces deux fonctions sur les deux exemples précédents

***Q2 corrigé***

L'implémentation de la première fonction est immédiate si l'on s'appuye sur l'initialisation $S_0=v_0$, et, pour tout $i \in [\![1, n]\!]$ sur la relation de récurrence, $S_i=S_{i-1} + v_i$.

```python
def som1(f, n):
    if n == 0 :
        return f(0)
    return som1(f, n - 1) + f(n)
```

* test

```python
som1(f, 9)
```

**Pour la seconde fonction, deux solutions récursives sont envisageables.**

La solution la plus immédiate s'appuie sur le fait que pour une liste `L` non vide la somme de ces termes est donnée par 
$$S_L = \sum_{k=0}^{{\rm len(L)}-1}{\rm L}[k]$$
et qu'en posant, pour tout $i \in [\![1, {\rm len(L)}-1]\!]$, $S_i=\sum_{k=0}^{i}{\rm L}[i]$, on a, en posant $S_{-1}=0$ :
$$
\forall i \in [\![0, {\rm len(L)}-1]\!], \quad S_{i} = S_{i - 1} + {\rm L}[i],
$$
ce qui peut se réécrire :
$$\left\{\begin{array}{l}
S_{[]} = 0\\
\forall i \in [\![0, {\rm len(L)}-1]\!], \quad S_{{\rm L}[0:i+1]} = S_{{\rm L}[0:i]} + {\rm L}[i].
\end{array}\right.$$

De façon informelle, on peut noter que la somme des termes d'une liste `L` est égale à zéro pour une liste vide, et, pour une liste non vide, égale à la somme de son dernier terme et de la somme des termes précédents (termes de la liste extraite `L[0:len(L)-1]`).

On en tire l'implémentation récursive suivante :

```python
def som2(L):
    if L == []:
        return 0
    return L[len(L) - 1] + som2(L[0:len(L) - 1])
```

* test

```python
som2([1, 2, 3, 4, 5])
```

De façon semblable, on peut noter la somme des termes d'une liste `L` est égale à zéro pour une liste vide, et, pour une liste non vide, égale à la somme de son <u>premier</u> terme et de la somme des termes suivants (termes de la liste extraite `L[1:len(L)]`).

On en tire l'implémentation récursive suivante :

```python
def som21(L):
    if L == []:
        return 0
    return L[0] + som2(L[1:len(L)])
```

***L'inconvénient de ces deux implémentations et que chaque appel entraîne la copie d'un liste (opération de <i>slicing</i>)***, ce qui a un coût en terme de temps de calcul :  pour une liste de $n$ termes, on copiera une liste de longueur $n-1$, une liste de longueur $n-2$, *etc.*, jusqu'à la copie d'une liste de longueur $1$, soit au total la copie de $(n-1) + (n-2) +\ldots+2+1=\frac{(n-1)n}{2}$ copies de termes.

**La solution suivante, bien que moins intuitive, <i>ne nécessite aucune copie de termes</i>, et est donc préférable.**

Cette troisième solution s'appuie sur le calcul par récurrence des sommes $S_i=\sum_{k=0}^{{\rm len(L)}-1}{\rm L}[k]$, avec  l'initialisation $S_0={\rm L}[0]$, et, pour tout $i \in [\![1, {\rm len(L)}-1]\!]$ sur la relation de récurrence, $S_i=S_{i-1} + {\rm L}[i]$.  
On obtient le calcul de la somme des termes d'une liste `L` en utilisant un second paramètre, $i$, initialisé à la valeur `len(L)-1`, la définition par récurrence devenant alors :
$$\left\{\begin{array}{l}
S_{-1} = 0\\
\forall i \in [\![0, {\rm len(L)}-1]\!], \quad S_{i} = S_{i - 1} + {\rm L}[i].
\end{array}\right.$$
Le cas terminal (ne nécessitant aucun appel récursif) sera alors pour $i=0$.

```python
def som22(L, i = len(L) - 1):
    if i == - 1 :
        return 0
    return L[i] + som22(L, i - 1)
```

```python
som22([1, 2, 3, 4, 5])
```

```python
def interclassement_base(L1, L2):
    L = []
    i1, i2 = 0, 0
    while i1 < len(L1) and i2 < len(L2):
        if L1[i1] <= L2[i2]:
            L.append(L1[i1])
            i1 += 1
        else:
            L.append(L2[i2])
            i2 += 1
    return L
```

**Q2** On donne la version suivante de la fonction `interclassement_base` :

```python
def interclassement(L1, L2):
    ## pré-traitement
    L1.append(float('inf'))
    L2.append(float('inf'))
    ##
    L = []
    i1, i2 = 0, 0
    while i1 < len(L1) and i2 < len(L2):
        if L1[i1] <= L2[i2]:
            L.append(L1[i1])
            i1 += 1
        else:
            L.append(L2[i2])
            i2 += 1
    ## annulation du pré-traitement
    L1.pop()
    L2.pop()
    ## post-traitement
    L.pop()
    return L
```
